```python
import matplotlib.pyplot as plt
import numpy as np

# Impostazioni iniziali
plt.rcParams['font.family'] = 'CMU Serif'
rounds = np.arange(1, 11)

# ------------------------------
# Dati per F1 Score (m1, m3_t1)
accuracy_m1    = [0.24, 0.24, 0.24, 0.25, 0.26, 0.26, 0.26, 0.28, 0.30, 0.33]#ok
accuracy_m3_t1 = [0.07, 0.11, 0.16, 0.2, 0.24, 0.28, 0.31, 0.34, 0.36, 0.38]

# ------------------------------
# Dati per MAE (m2, m3_t2)
# Generiamo 10 valori casuali per ciascun modello (MAE compreso tra 0.1 e 0.5)
np.random.seed(0)
mae_m2    = np.random.uniform(0.1, 0.5, 10)
mae_m3_t2 = np.random.uniform(0.1, 0.5, 10)

# ------------------------------
# Dati per Average Training Time
# (Si utilizzano m1, m2 e m3_t1; m3_t2 non è previsto in questo grafico)
total_training_time_m1    = [444, 398, 445, 471, 467, 407, 467, 470, 495, 595]#ok
total_training_time_m2    = [110, 115, 120, 125, 130, 135, 140, 145, 150, 155]
total_training_time_m3_t1 = [95, 100, 105, 110, 115, 120, 125, 130, 135, 140]
total_training_time_m3_t2 = [95, 100, 105, 110, 115, 120, 125, 130, 135, 140]

# Calcolo dei rapporti
f1_over_training_m1    = [a / t for a, t in zip(accuracy_m1, total_training_time_m1)]
f1_over_training_m3_t1 = [a / t for a, t in zip(accuracy_m3_t1, total_training_time_m3_t1)]
mae_over_training_m2    = [m / t for m, t in zip(mae_m2, total_training_time_m2)]
mae_over_training_m3_t2 = [m / t for m, t in zip(mae_m3_t2, total_training_time_m3_t2)]

# ------------------------------
# Grafico 1: F1 Score vs FL Round (m1 e m3_t1)
colors1  = ['#66c2a5', '#fc8d62']
labels1  = [r'Config. $T_{1}$ ($M_{1}$)', r'Config. $T_{2} || T_{3}$ ($M_{3}$)']
markers1 = ['o', '^']

fig, ax = plt.subplots(figsize=(14, 10))
for data, label, color, marker in zip([accuracy_m1, accuracy_m3_t1], labels1, colors1, markers1):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('F1 Score', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper center', bbox_to_anchor=(0.5, 1.12), ncol=3)
plt.tight_layout()
plt.savefig('F1_Score.pdf', bbox_inches='tight')
plt.show()

# ------------------------------
# Grafico 2: MAE vs FL Round (m2 e m3_t2)
colors2  = ['#1b9e77', '#e34a33']
labels2  = [r'Config. $T_{2}$ ($M_{2}$)', r'Config. $T_{2} || T_{3}$ ($M_{3}$)']
markers2 = ['s', 'd']

fig, ax = plt.subplots(figsize=(14, 10))
for data, label, color, marker in zip([mae_m2, mae_m3_t2], labels2, colors2, markers2):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('Mean Absolute Error (m)', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper center', bbox_to_anchor=(0.5, 1.12), ncol=2)
plt.tight_layout()
plt.savefig('MAE.pdf', bbox_inches='tight')
plt.show()

# ------------------------------
# Grafico 3: Average Training Time vs FL Round (m1 e m3)
labels_at1 = [r'Config. $T_{1}$ ($M_{1}$)', r'Config. $T_{2} || T_{3}$ ($M_{3}$)']
colors_at1 = ['#66c2a5', '#fc8d62']
markers_at1 = ['o', '^']

fig, ax = plt.subplots(figsize=(14, 10))
for data, label, color, marker in zip([total_training_time_m1, total_training_time_m3_t1], labels_at1, colors_at1, markers_at1):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('Average Training Time (s)', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper center', bbox_to_anchor=(0.5, 1.12), ncol=2)
plt.tight_layout()
plt.savefig('TT_M1_M3.pdf', bbox_inches='tight')
plt.show()

# ------------------------------
# Grafico 4: Average Training Time vs FL Round (m2 e m3)
labels_at2 = [r'Config. $T_{2}$ ($M_{2}$)', r'Config. $T_{2} || T_{3}$ ($M_{3}$)']
colors_at2 = ['#1b9e77', '#e34a33']
markers_at2 = ['s', 'd']

fig, ax = plt.subplots(figsize=(14, 10))
for data, label, color, marker in zip([total_training_time_m2, total_training_time_m3_t1], labels_at2, colors_at2, markers_at2):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('Average Training Time (s)', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper center', bbox_to_anchor=(0.5, 1.12), ncol=2)
plt.tight_layout()
plt.savefig('TT_M2_M3.pdf', bbox_inches='tight')
plt.show()

# ------------------------------
# Grafico 5: F1 Score / Average Training Time (m1 e m3_t1)
fig, ax = plt.subplots(figsize=(14, 10))
for data, label, color, marker in zip([f1_over_training_m1, f1_over_training_m3_t1], labels1, colors1, markers1):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('F1 Score / Average Training Time', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper center', bbox_to_anchor=(0.5, 1.12), ncol=2)
plt.tight_layout()
plt.savefig('Efficiency_M1_M3.pdf', bbox_inches='tight')
plt.show()

# ------------------------------
# Grafico 6: MAE / Average Training Time (m2 e m3_t2)
fig, ax = plt.subplots(figsize=(14, 10))
for data, label, color, marker in zip([mae_over_training_m2, mae_over_training_m3_t2], labels2, colors2, markers2):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('MAE / Average Training Time', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper center', bbox_to_anchor=(0.5, 1.12), ncol=2)
plt.tight_layout()
plt.savefig('Efficiency_M2_M3.pdf', bbox_inches='tight')
plt.show()
```
