```python
import matplotlib.pyplot as plt
import numpy as np

# Define the rounds
rounds = np.arange(1, 11)
plt.rcParams['font.family'] = 'CMU Serif'

# Randomly generated accuracy data
accuracy_A = [0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.38, 0.4, 0.42, 0.45]
accuracy_B = [0.08, 0.12, 0.18, 0.22, 0.26, 0.3, 0.33, 0.36, 0.38, 0.4]
accuracy_C = [0.07, 0.11, 0.16, 0.2, 0.24, 0.28, 0.31, 0.34, 0.36, 0.38]
accuracy_D = [0.09, 0.13, 0.17, 0.21, 0.25, 0.29, 0.32, 0.35, 0.37, 0.39]

# Randomly generated total round time data
total_round_time_A = [200, 210, 220, 230, 240, 250, 260, 270, 280, 290]
total_round_time_B = [210, 220, 230, 240, 250, 260, 270, 280, 290, 300]
total_round_time_C = [190, 200, 210, 220, 230, 240, 250, 260, 270, 280]
total_round_time_D = [205, 215, 225, 235, 245, 255, 265, 275, 285, 295]

# Randomly generated average training time data
total_training_time_A = [100, 105, 110, 115, 120, 125, 130, 135, 140, 145]
total_training_time_B = [110, 115, 120, 125, 130, 135, 140, 145, 150, 155]
total_training_time_C = [95, 100, 105, 110, 115, 120, 125, 130, 135, 140]
total_training_time_D = [105, 110, 115, 120, 125, 130, 135, 140, 145, 150]

# Calculate Accuracy over Total Round Time and multiply by 1000
accuracy_over_time_A = [1000 * (a / t) for a, t in zip(accuracy_A, total_round_time_A)]
accuracy_over_time_B = [1000 * (a / t) for a, t in zip(accuracy_B, total_round_time_B)]
accuracy_over_time_C = [1000 * (a / t) for a, t in zip(accuracy_C, total_round_time_C)]
accuracy_over_time_D = [1000 * (a / t) for a, t in zip(accuracy_D, total_round_time_D)]

# Calculate Accuracy over Training Time and multiply by 1000
accuracy_over_training_time_A = [1000 * (a / t) for a, t in zip(accuracy_A, total_training_time_A)]
accuracy_over_training_time_B = [1000 * (a / t) for a, t in zip(accuracy_B, total_training_time_B)]
accuracy_over_training_time_C = [1000 * (a / t) for a, t in zip(accuracy_C, total_training_time_C)]
accuracy_over_training_time_D = [1000 * (a / t) for a, t in zip(accuracy_D, total_training_time_D)]

# Colors and styles
colors = ['#66c2a5', '#1b9e77', '#fc8d62', '#e34a33']
labels = ['Config. A', 'Config. B', 'Config. C', 'Config. D']
markers = ['o', 's', '^', 'd']

### First Plot: F1 Score ###
fig1, ax1 = plt.subplots(figsize=(14, 10))

# Plot the curves
for i, (data, label, color, marker) in enumerate(zip([accuracy_A, accuracy_B, accuracy_C, accuracy_D], labels, colors, markers)):
    ax1.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')

ax1.set_xlabel('Federated Learning Round', fontsize=30)
ax1.set_ylabel('F1 Score', fontsize=30)
ax1.tick_params(axis='both', labelsize=30)
ax1.set_xticks(rounds)
ax1.legend(fontsize=25, loc='upper left')
plt.tight_layout()
plt.savefig('Configuration_F1_Score.pdf', bbox_inches='tight')
plt.show()

### Second Plot: Total Round Time ###
fig2, ax2 = plt.subplots(figsize=(14, 10))

for i, (data, label, color, marker) in enumerate(zip([total_round_time_A, total_round_time_B, total_round_time_C, total_round_time_D], labels, colors, markers)):
    ax2.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')

ax2.set_xlabel('Federated Learning Round', fontsize=30)
ax2.set_ylabel('Total Round Time (s)', fontsize=30)
ax2.tick_params(axis='both', labelsize=30)
ax2.set_xticks(rounds)
ax2.legend(fontsize=25, loc='upper left')
plt.tight_layout()
plt.savefig('Configuration_Total_Round_Time.pdf', bbox_inches='tight')
plt.show()

### Third Plot: Accuracy / Total Round Time ###
fig3, ax3 = plt.subplots(figsize=(14, 10))

for i, (data, label, color, marker) in enumerate(zip([accuracy_over_time_A, accuracy_over_time_B, accuracy_over_time_C, accuracy_over_time_D], labels, colors, markers)):
    ax3.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')

ax3.set_xlabel('Federated Learning Round', fontsize=30)
ax3.set_ylabel('F1 Score / Total Round Time', fontsize=30)
ax3.tick_params(axis='both', labelsize=30)
ax3.set_xticks(rounds)
ax3.legend(fontsize=25, loc='upper left')
plt.tight_layout()
plt.savefig('Configuration_Accuracy_Over_Round_Time.pdf', bbox_inches='tight')
plt.show()

### Fourth Plot: Average Training Time ###
fig4, ax4 = plt.subplots(figsize=(14, 10))

for i, (data, label, color, marker) in enumerate(zip([total_training_time_A, total_training_time_B, total_training_time_C, total_training_time_D], labels, colors, markers)):
    ax4.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')

ax4.set_xlabel('Federated Learning Round', fontsize=30)
ax4.set_ylabel('Average Training Time (s)', fontsize=30)
ax4.tick_params(axis='both', labelsize=30)
ax4.set_xticks(rounds)
ax4.legend(fontsize=25, loc='upper left')
plt.tight_layout()
plt.savefig('Configuration_Average_Training_Time.pdf', bbox_inches='tight')
plt.show()

### Fifth Plot: Accuracy / Average Training Time ###
fig5, ax5 = plt.subplots(figsize=(14, 10))

for i, (data, label, color, marker) in enumerate(zip([accuracy_over_training_time_A, accuracy_over_training_time_B, accuracy_over_training_time_C, accuracy_over_training_time_D], labels, colors, markers)):
    ax5.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')

ax5.set_xlabel('Federated Learning Round', fontsize=30)
ax5.set_ylabel('F1 Score / Average Training Time', fontsize=30)
ax5.tick_params(axis='both', labelsize=30)
ax5.set_xticks(rounds)
ax5.legend(fontsize=25, loc='upper left')
plt.tight_layout()
plt.savefig('Configuration_Accuracy_Over_Training_Time.pdf', bbox_inches='tight')
plt.show()
```
