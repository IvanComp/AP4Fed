```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches  

# Define the rounds
rounds = np.arange(1, 11)
plt.rcParams['font.family'] = 'CMU Serif'

# Accuracy data provided
accuracy_4a4b_cluster = [0.12, 0.22, 0.29, 0.35, 0.38, 0.41, 0.44, 0.46, 0.47, 0.48]
accuracy_4a4b_no_cluster = [0.11, 0.12, 0.15, 0.17, 0.20, 0.19, 0.22, 0.21, 0.21, 0.22]
accuracy_5a3b_cluster = [0.08, 0.11, 0.23, 0.30, 0.35, 0.39, 0.40, 0.42, 0.44, 0.45]
accuracy_5a3b_no_cluster = [0.08, 0.07, 0.08, 0.13, 0.17, 0.16, 0.19, 0.17, 0.17, 0.19]
accuracy_6a2b_cluster = [0.06, 0.16, 0.23, 0.29, 0.34, 0.37, 0.40, 0.40, 0.44, 0.44]
accuracy_6a2b_no_cluster = [0.09, 0.09, 0.12, 0.14, 0.16, 0.17, 0.19, 0.19, 0.23, 0.21]

# Total Training Time (average) data
total_training_time_4a4b_cluster = [109, 110, 121, 107, 118, 104, 106, 111, 105, 101]
total_training_time_4a4b_no_cluster = [189, 200, 199, 183, 175, 170, 182, 175, 166, 168]
total_training_time_5a3b_cluster = [112, 110, 114, 110, 118, 112, 110, 116, 114, 113]
total_training_time_5a3b_no_cluster = [209, 222, 215, 230, 220, 235, 220, 220, 225, 209]
total_training_time_6a2b_cluster = [138, 127, 128, 134, 136, 123, 121, 127, 128, 130]
total_training_time_6a2b_no_cluster = [170, 160, 158, 153, 162, 158, 172, 160, 161, 166]

# Calculate Accuracy over Training Time and multiply by 1000
accuracy_over_training_time_4a4b_cluster = [1000 * (a / t) for a, t in zip(accuracy_4a4b_cluster, total_training_time_4a4b_cluster)]
accuracy_over_training_time_4a4b_no_cluster = [1000 * (a / t) for a, t in zip(accuracy_4a4b_no_cluster, total_training_time_4a4b_no_cluster)]
accuracy_over_training_time_5a3b_cluster = [1000 * (a / t) for a, t in zip(accuracy_5a3b_cluster, total_training_time_5a3b_cluster)]
accuracy_over_training_time_5a3b_no_cluster = [1000 * (a / t) for a, t in zip(accuracy_5a3b_no_cluster, total_training_time_5a3b_no_cluster)]
accuracy_over_training_time_6a2b_cluster = [1000 * (a / t) for a, t in zip(accuracy_6a2b_cluster, total_training_time_6a2b_cluster)]
accuracy_over_training_time_6a2b_no_cluster = [1000 * (a / t) for a, t in zip(accuracy_6a2b_no_cluster, total_training_time_6a2b_no_cluster)]

# Colors and styles
color_4a4b_cluster = '#66c2a5'   # Light teal for 4a4b with cluster
color_5a3b_cluster = '#1b9e77'   # Medium teal for 5a3b with cluster
color_6a2b_cluster = '#01665e'   # Dark teal for 6a2b with cluster
color_4a4b_no_cluster = '#fc8d62'  # Light coral for 4a4b no cluster
color_5a3b_no_cluster = '#e34a33'  # Medium coral for 5a3b no cluster
color_6a2b_no_cluster = '#b2182b'  # Dark coral for 6a2b no cluster

# Variability for confidence intervals (artificial standard deviation)
std_accuracy = 0.02 # example value for accuracy
std_time = 10 # example value for training time

# Calculate Confidence Intervals for Accuracy and Training Time
accuracy_conf_4a4b_cluster_upper = np.array(accuracy_4a4b_cluster) + std_accuracy
accuracy_conf_4a4b_cluster_lower = np.array(accuracy_4a4b_cluster) - std_accuracy
accuracy_conf_4a4b_no_cluster_upper = np.array(accuracy_4a4b_no_cluster) + std_accuracy
accuracy_conf_4a4b_no_cluster_lower = np.array(accuracy_4a4b_no_cluster) - std_accuracy
accuracy_conf_5a3b_cluster_upper = np.array(accuracy_5a3b_cluster) + std_accuracy
accuracy_conf_5a3b_cluster_lower = np.array(accuracy_5a3b_cluster) - std_accuracy
accuracy_conf_5a3b_no_cluster_upper = np.array(accuracy_5a3b_no_cluster) + std_accuracy
accuracy_conf_5a3b_no_cluster_lower = np.array(accuracy_5a3b_no_cluster) - std_accuracy
accuracy_conf_6a2b_cluster_upper = np.array(accuracy_6a2b_cluster) + std_accuracy
accuracy_conf_6a2b_cluster_lower = np.array(accuracy_6a2b_cluster) - std_accuracy
accuracy_conf_6a2b_no_cluster_upper = np.array(accuracy_6a2b_no_cluster) + std_accuracy
accuracy_conf_6a2b_no_cluster_lower = np.array(accuracy_6a2b_no_cluster) - std_accuracy

# Create plots with shaded confidence intervals

fig1, ax1 = plt.subplots(figsize=(14, 10))

# Plot the curves and confidence intervals
ax1.plot(rounds, accuracy_4a4b_cluster, color=color_4a4b_cluster, label='4a4b with cluster',
         marker='o', linewidth=2, linestyle='--')
ax1.fill_between(rounds, accuracy_conf_4a4b_cluster_lower, accuracy_conf_4a4b_cluster_upper,
                  color=color_4a4b_cluster, alpha=0.2)

ax1.plot(rounds, accuracy_4a4b_no_cluster, color=color_4a4b_no_cluster, label='4a4b no cluster',
         marker='s', linewidth=2, linestyle='-')
ax1.fill_between(rounds, accuracy_conf_4a4b_no_cluster_lower, accuracy_conf_4a4b_no_cluster_upper,
                  color=color_4a4b_no_cluster, alpha=0.2)

ax1.plot(rounds, accuracy_5a3b_cluster, color=color_5a3b_cluster, label='5a3b with cluster',
         marker='^', linewidth=2, linestyle='--')
ax1.fill_between(rounds, accuracy_conf_5a3b_cluster_lower, accuracy_conf_5a3b_cluster_upper,
                  color=color_5a3b_cluster, alpha=0.2)

ax1.plot(rounds, accuracy_5a3b_no_cluster, color=color_5a3b_no_cluster, label='5a3b no cluster',
         marker='d', linewidth=2, linestyle='-')
ax1.fill_between(rounds, accuracy_conf_5a3b_no_cluster_lower, accuracy_conf_5a3b_no_cluster_upper,
                  color=color_5a3b_no_cluster, alpha=0.2)

ax1.plot(rounds, accuracy_6a2b_cluster, color=color_6a2b_cluster, label='6a2b with cluster',
         marker='*', linewidth=2, linestyle='--')
ax1.fill_between(rounds, accuracy_conf_6a2b_cluster_lower, accuracy_conf_6a2b_cluster_upper,
                  color=color_6a2b_cluster, alpha=0.2)

ax1.plot(rounds, accuracy_6a2b_no_cluster, color=color_6a2b_no_cluster, label='6a2b no cluster',
         marker='p', linewidth=2, linestyle='-')
ax1.fill_between(rounds, accuracy_conf_6a2b_no_cluster_lower, accuracy_conf_6a2b_no_cluster_upper,
                  color=color_6a2b_no_cluster, alpha=0.2)

# Labels, legend, and axis settings
ax1.set_xlabel('Federated Learning Round', fontsize=30)
ax1.set_ylabel('F1 Score for Model A', fontsize=30)
ax1.tick_params(axis='both', labelsize=30)
ax1.legend(fontsize=25, loc='upper center', bbox_to_anchor=(0.5, 1.19), ncol=3)
ax1.set_xticks(rounds)
ax1.set_ylim(0, 0.5)

plt.tight_layout()
plt.savefig('ClientClustering1.pdf', bbox_inches='tight')
plt.show()

### Fourth Plot: Average Training Time ###
fig4, ax4 = plt.subplots(figsize=(14, 10))

# Plot average training time with confidence intervals
ax4.plot(rounds, total_training_time_4a4b_cluster, color=color_4a4b_cluster, label='4a4b with cluster',
         marker='o', linewidth=2, linestyle='--')
ax4.fill_between(rounds, np.array(total_training_time_4a4b_cluster) - std_time, 
                  np.array(total_training_time_4a4b_cluster) + std_time, color=color_4a4b_cluster, alpha=0.2)

ax4.plot(rounds, total_training_time_4a4b_no_cluster, color=color_4a4b_no_cluster, label='4a4b no cluster',
         marker='s', linewidth=2, linestyle='-')
ax4.fill_between(rounds, np.array(total_training_time_4a4b_no_cluster) - std_time, 
                  np.array(total_training_time_4a4b_no_cluster) + std_time, color=color_4a4b_no_cluster, alpha=0.2)

ax4.plot(rounds, total_training_time_5a3b_cluster, color=color_5a3b_cluster, label='5a3b with cluster',
         marker='^', linewidth=2, linestyle='--')
ax4.fill_between(rounds, np.array(total_training_time_5a3b_cluster) - std_time, 
                  np.array(total_training_time_5a3b_cluster) + std_time, color=color_5a3b_cluster, alpha=0.2)

ax4.plot(rounds, total_training_time_5a3b_no_cluster, color=color_5a3b_no_cluster, label='5a3b no cluster',
         marker='d', linewidth=2, linestyle='-')
ax4.fill_between(rounds, np.array(total_training_time_5a3b_no_cluster) - std_time, 
                  np.array(total_training_time_5a3b_no_cluster) + std_time, color=color_5a3b_no_cluster, alpha=0.2)

ax4.plot(rounds, total_training_time_6a2b_cluster, color=color_6a2b_cluster, label='6a2b with cluster',
         marker='*', linewidth=2, linestyle='--')
ax4.fill_between(rounds, np.array(total_training_time_6a2b_cluster) - std_time, 
                  np.array(total_training_time_6a2b_cluster) + std_time, color=color_6a2b_cluster, alpha=0.2)

ax4.plot(rounds, total_training_time_6a2b_no_cluster, color=color_6a2b_no_cluster, label='6a2b no cluster',
         marker='p', linewidth=2, linestyle='-')
ax4.fill_between(rounds, np.array(total_training_time_6a2b_no_cluster) - std_time, 
                  np.array(total_training_time_6a2b_no_cluster) + std_time, color=color_6a2b_no_cluster, alpha=0.2)

# Labels, legend, and axis settings
ax4.set_xlabel('Federated Learning Round', fontsize=30)
ax4.set_ylabel('Average Training Time (s)', fontsize=30)
ax4.tick_params(axis='both', labelsize=30)
ax4.legend(fontsize=25, loc='upper center', bbox_to_anchor=(0.5, 1.18), ncol=3)
plt.tight_layout()
ax4.set_xticks(rounds)
plt.savefig('ClientClustering2.pdf', bbox_inches='tight')
plt.show()

# Propagazione dell'errore per accuracy_over_training_time
def propagate_error(accuracy, time, std_acc, std_time):
    error_relative = np.sqrt((std_acc / np.array(accuracy))**2 + 
                             (std_time / np.array(time))**2)
    error_absolute = np.array([1000 * (a / t) for a, t in zip(accuracy, time)]) * error_relative
    return error_absolute

# Calcolo degli errori assoluti
error_4a4b_cluster = propagate_error(accuracy_4a4b_cluster, total_training_time_4a4b_cluster, std_accuracy, std_time)
error_4a4b_no_cluster = propagate_error(accuracy_4a4b_no_cluster, total_training_time_4a4b_no_cluster, std_accuracy, std_time)
error_5a3b_cluster = propagate_error(accuracy_5a3b_cluster, total_training_time_5a3b_cluster, std_accuracy, std_time)
error_5a3b_no_cluster = propagate_error(accuracy_5a3b_no_cluster, total_training_time_5a3b_no_cluster, std_accuracy, std_time)
error_6a2b_cluster = propagate_error(accuracy_6a2b_cluster, total_training_time_6a2b_cluster, std_accuracy, std_time)
error_6a2b_no_cluster = propagate_error(accuracy_6a2b_no_cluster, total_training_time_6a2b_no_cluster, std_accuracy, std_time)


### Fifth Plot: Accuracy / Average Training Time ###
fig5, ax5 = plt.subplots(figsize=(14, 10))

# Aggiungi bande di confidenza propagate per ogni curva
ax5.plot(rounds, accuracy_over_training_time_4a4b_cluster, color=color_4a4b_cluster, label='4a4b with cluster',
         marker='o', linewidth=2, linestyle='--')
ax5.fill_between(rounds, 
                 np.array(accuracy_over_training_time_4a4b_cluster) - error_4a4b_cluster, 
                 np.array(accuracy_over_training_time_4a4b_cluster) + error_4a4b_cluster, 
                 color=color_4a4b_cluster, alpha=0.2)

ax5.plot(rounds, accuracy_over_training_time_4a4b_no_cluster, color=color_4a4b_no_cluster, label='4a4b no cluster',
         marker='s', linewidth=2, linestyle='-')
ax5.fill_between(rounds, 
                 np.array(accuracy_over_training_time_4a4b_no_cluster) - error_4a4b_no_cluster, 
                 np.array(accuracy_over_training_time_4a4b_no_cluster) + error_4a4b_no_cluster, 
                 color=color_4a4b_no_cluster, alpha=0.2)

ax5.plot(rounds, accuracy_over_training_time_5a3b_cluster, color=color_5a3b_cluster, label='5a3b with cluster',
         marker='^', linewidth=2, linestyle='--')
ax5.fill_between(rounds, 
                 np.array(accuracy_over_training_time_5a3b_cluster) - error_5a3b_cluster, 
                 np.array(accuracy_over_training_time_5a3b_cluster) + error_5a3b_cluster, 
                 color=color_5a3b_cluster, alpha=0.2)

ax5.plot(rounds, accuracy_over_training_time_5a3b_no_cluster, color=color_5a3b_no_cluster, label='5a3b no cluster',
         marker='d', linewidth=2, linestyle='-')
ax5.fill_between(rounds, 
                 np.array(accuracy_over_training_time_5a3b_no_cluster) - error_5a3b_no_cluster, 
                 np.array(accuracy_over_training_time_5a3b_no_cluster) + error_5a3b_no_cluster, 
                 color=color_5a3b_no_cluster, alpha=0.2)

ax5.plot(rounds, accuracy_over_training_time_6a2b_cluster, color=color_6a2b_cluster, label='6a2b with cluster',
         marker='*', linewidth=2, linestyle='--')
ax5.fill_between(rounds, 
                 np.array(accuracy_over_training_time_6a2b_cluster) - error_6a2b_cluster, 
                 np.array(accuracy_over_training_time_6a2b_cluster) + error_6a2b_cluster, 
                 color=color_6a2b_cluster, alpha=0.2)

ax5.plot(rounds, accuracy_over_training_time_6a2b_no_cluster, color=color_6a2b_no_cluster, label='6a2b no cluster',
         marker='p', linewidth=2, linestyle='-')
ax5.fill_between(rounds, 
                 np.array(accuracy_over_training_time_6a2b_no_cluster) - error_6a2b_no_cluster, 
                 np.array(accuracy_over_training_time_6a2b_no_cluster) + error_6a2b_no_cluster, 
                 color=color_6a2b_no_cluster, alpha=0.2)

# Labels, legend, and axis settings
ax5.set_xlabel('Federated Learning Round', fontsize=30)
ax5.set_ylabel('F1 Score / Average Training Time', fontsize=30)
ax5.tick_params(axis='both', labelsize=30)
ax5.legend(fontsize=25, loc='upper center', bbox_to_anchor=(0.5, 1.19), ncol=3)
ax5.set_xticks(rounds)

plt.tight_layout()
plt.savefig('ClientClustering3.pdf', bbox_inches='tight')
plt.show()
```
