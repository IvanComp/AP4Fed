```python
import matplotlib.pyplot as plt
import numpy as np

# Impostazioni iniziali
plt.rcParams['font.family'] = 'CMU Serif'
rounds = np.arange(1, 11)

# ------------------------------
# Dati per F1 Score (m1, m3_t1)
accuracy_m1    = [0.1,  0.15, 0.2,  0.25, 0.3,  0.35, 0.38, 0.4,  0.42, 0.45]
accuracy_m3_t1 = [0.07, 0.11, 0.16, 0.2,  0.24, 0.28, 0.31, 0.34, 0.36, 0.38]

# ------------------------------
# Dati per MAE (m2, m3_t2)
# Generiamo 10 valori casuali per ciascun modello (MAE compreso tra 0.1 e 0.5)
np.random.seed(0)
mae_m2    = np.random.uniform(0.1, 0.5, 10)
mae_m3_t2 = np.random.uniform(0.1, 0.5, 10)

# ------------------------------
# Dati per Total Round Time
# (Si utilizzano m1, m2 e m3_t1; m3_t2 non è previsto in questo grafico)
total_round_time_m1    = [200, 210, 220, 230, 240, 250, 260, 270, 280, 290]
total_round_time_m2    = [210, 220, 230, 240, 250, 260, 270, 280, 290, 300]
total_round_time_m3_t1 = [190, 200, 210, 220, 230, 240, 250, 260, 270, 280]

# ------------------------------
# Dati per Average Training Time
# (Si utilizzano m1, m2 e m3_t1; m3_t2 non è previsto in questo grafico)
total_training_time_m1    = [100, 105, 110, 115, 120, 125, 130, 135, 140, 145]
total_training_time_m2    = [110, 115, 120, 125, 130, 135, 140, 145, 150, 155]
total_training_time_m3_t1 = [95,  100, 105, 110, 115, 120, 125, 130, 135, 140]

# ------------------------------
# Grafico 1: F1 Score vs FL Round (m1 e m3_t1)
colors1  = ['#66c2a5', '#fc8d62']
labels1  = [r'$M_{1}$', r'$M_{3_{\text{(t1)}}}$']
markers1 = ['o', '^']

fig, ax = plt.subplots(figsize=(14, 10))
for data, label, color, marker in zip([accuracy_m1, accuracy_m3_t1], labels1, colors1, markers1):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('F1 Score', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper left')
plt.tight_layout()
plt.savefig('F1_Score.pdf', bbox_inches='tight')
plt.show()

# ------------------------------
# Grafico 2: MAE vs FL Round (m2 e m3_t2)
colors2  = ['#1b9e77', '#e34a33']
labels2  = [r'$M_{2}$', r'$M_{3_{\text{(t2)}}}$']
markers2 = ['s', 'd']

fig, ax = plt.subplots(figsize=(14, 10))
for data, label, color, marker in zip([mae_m2, mae_m3_t2], labels2, colors2, markers2):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('Mean Absolute Error (m)', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper left')
plt.tight_layout()
plt.savefig('MAE.pdf', bbox_inches='tight')
plt.show()

# ------------------------------
# Grafico 3: Total Round Time vs FL Round (m1, m2, m3_t1)
colors3  = ['#66c2a5', '#1b9e77', '#fc8d62']
labels3  = [r'$M_{1}$', r'$M_{2}$', r'$M_{3}$']
markers3 = ['o', 's', '^']

fig, ax = plt.subplots(figsize=(14, 10))
for data, label, color, marker in zip([total_round_time_m1, total_round_time_m2, total_round_time_m3_t1], labels3, colors3, markers3):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('Total Round Time (s)', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper left')
plt.tight_layout()
plt.savefig('Total_Round_Time.pdf', bbox_inches='tight')
plt.show()

# ------------------------------
# Grafico 4: Average Training Time vs FL Round (m1, m2, m3_t1)
fig, ax = plt.subplots(figsize=(14, 10))
for data, label, color, marker in zip([total_training_time_m1, total_training_time_m2, total_training_time_m3_t1], labels3, colors3, markers3):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('Average Training Time (s)', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper left')
plt.tight_layout()
plt.savefig('Average_Training_Time.pdf', bbox_inches='tight')
plt.show()

# ------------------------------
# Calcolo dei rapporti per F1 Score (solo m1 e m3_t1)
# F1 Score / Total Round Time
f1_over_total_m1    = [a / t for a, t in zip(accuracy_m1,    total_round_time_m1)]
f1_over_total_m3_t1 = [a / t for a, t in zip(accuracy_m3_t1, total_round_time_m3_t1)]
# F1 Score / Average Training Time
f1_over_training_m1    = [a / t for a, t in zip(accuracy_m1,    total_training_time_m1)]
f1_over_training_m3_t1 = [a / t for a, t in zip(accuracy_m3_t1, total_training_time_m3_t1)]

# Grafico 5a: F1 Score / Total Round Time (m1 e m3_t1)
fig, ax = plt.subplots(figsize=(14, 10))
for data, label, color, marker in zip([f1_over_total_m1, f1_over_total_m3_t1], labels1, colors1, markers1):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('F1 Score / Total Round Time', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper left')
plt.tight_layout()
plt.savefig('F1_over_Total_Round_Time.pdf', bbox_inches='tight')
plt.show()

# Grafico 5b: F1 Score / Average Training Time (m1 e m3_t1)
fig, ax = plt.subplots(figsize=(14, 10))
for data, label, color, marker in zip([f1_over_training_m1, f1_over_training_m3_t1], labels1, colors1, markers1):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('F1 Score / Average Training Time', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper left')
plt.tight_layout()
plt.savefig('F1_over_Average_Training_Time.pdf', bbox_inches='tight')
plt.show()

# ------------------------------
# Calcolo dei rapporti per MAE (m2 e m3_t2)
# MAE / Total Round Time
mae_over_total_m2    = [m / t for m, t in zip(mae_m2,    total_round_time_m2)]
mae_over_total_m3_t2 = [m / t for m, t in zip(mae_m3_t2, total_round_time_m3_t2)]
# MAE / Average Training Time
mae_over_training_m2    = [m / t for m, t in zip(mae_m2,    total_training_time_m2)]
mae_over_training_m3_t2 = [m / t for m, t in zip(mae_m3_t2, total_training_time_m3_t2)]

# Grafico 6a: MAE / Total Round Time (m2 e m3_t2)
fig, ax = plt.subplots(figsize=(14, 10))
for data, label, color, marker in zip([mae_over_total_m2, mae_over_total_m3_t2], labels2, colors2, markers2):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('MAE / Total Round Time', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper left')
plt.tight_layout()
plt.savefig('MAE_over_Total_Round_Time.pdf', bbox_inches='tight')
plt.show()

# Grafico 6b: MAE / Average Training Time (m2 e m3_t2)
fig, ax = plt.subplots(figsize=(14, 10))
for data, label, color, marker in zip([mae_over_training_m2, mae_over_training_m3_t2], labels2, colors2, markers2):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('MAE / Average Training Time', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper left')
plt.tight_layout()
plt.savefig('MAE_over_Average_Training_Time.pdf', bbox_inches='tight')
plt.show()
```
