```python
import matplotlib.pyplot as plt
import numpy as np

# Impostazioni iniziali
plt.rcParams['font.family'] = 'CMU Serif'
rounds = np.arange(1, 11)

# Define consistent colors for all plots
COLOR_M1 = "#1f77b4"  # blue for m1
COLOR_M2 = "#d62728"  # red for m2
COLOR_M3 = "#2ca02c"  # green for m3

# ------------------------------
# Dati per F1 Score (m1, m3_t1)
accuracy_m1    = [0.24, 0.24, 0.24, 0.25, 0.26, 0.26, 0.26, 0.28, 0.30, 0.33]
accuracy_m3_t1 = [0.07, 0.11, 0.16, 0.2, 0.24, 0.28, 0.31, 0.34, 0.36, 0.38]

# Confidence intervals for F1 Score
ci_m1_f1 = [0.005, 0.0048, 0.0052, 0.0051, 0.0049, 0.0053, 0.0047, 0.005, 0.0048, 0.0052]
ci_m3_t1_f1 = [0.002, 0.003, 0.0035, 0.004, 0.0045, 0.005, 0.0048, 0.005, 0.0052, 0.0055]

# ------------------------------
# Dati per MAE (m2, m3_t2)
np.random.seed(0)
mae_m2    = np.random.uniform(0.1, 0.5, 10)
mae_m3_t2 = np.random.uniform(0.1, 0.5, 10)

# Confidence intervals for MAE
ci_m2_mae = [0.015, 0.018, 0.016, 0.019, 0.017, 0.02, 0.018, 0.021, 0.019, 0.022]
ci_m3_t2_mae = [0.012, 0.015, 0.014, 0.016, 0.015, 0.017, 0.016, 0.018, 0.017, 0.019]

# ------------------------------
# Dati per Average Training Time
total_training_time_m1    = [444, 398, 445, 471, 467, 407, 467, 470, 495, 595]
total_training_time_m2    = [110, 115, 120, 125, 130, 135, 140, 145, 150, 155]
total_training_time_m3_t1 = [95, 100, 105, 110, 115, 120, 125, 130, 135, 140]
total_training_time_m3_t2 = [95, 100, 105, 110, 115, 120, 125, 130, 135, 140]

# Confidence intervals for Training Time
ci_m1_tt = [8, 7, 8, 9, 8, 7, 9, 8, 10, 12]
ci_m2_tt = [5, 6, 5, 6, 7, 6, 7, 6, 7, 8]
ci_m3_tt = [4, 5, 4, 5, 5, 6, 5, 6, 6, 7]

# Calcolo dei rapporti
f1_over_training_m1    = [a / t for a, t in zip(accuracy_m1, total_training_time_m1)]
f1_over_training_m3_t1 = [a / t for a, t in zip(accuracy_m3_t1, total_training_time_m3_t1)]
mae_over_training_m2    = [m / t for m, t in zip(mae_m2, total_training_time_m2)]
mae_over_training_m3_t2 = [m / t for m, t in zip(mae_m3_t2, total_training_time_m3_t2)]

# Confidence intervals for efficiency metrics
ci_eff_m1 = [0.00012, 0.00013, 0.00012, 0.00014, 0.00013, 0.00015, 0.00014, 0.00015, 0.00015, 0.00016]
ci_eff_m3_t1 = [0.00008, 0.00009, 0.0001, 0.00011, 0.00012, 0.00013, 0.00014, 0.00015, 0.00015, 0.00016]
ci_eff_m2 = [0.0002, 0.00022, 0.00021, 0.00023, 0.00022, 0.00024, 0.00023, 0.00025, 0.00024, 0.00026]
ci_eff_m3_t2 = [0.00015, 0.00016, 0.00016, 0.00017, 0.00017, 0.00018, 0.00018, 0.00019, 0.00019, 0.0002]

# ------------------------------
# Grafico 1: F1 Score vs FL Round (m1 e m3_t1)
colors1  = [COLOR_M1, COLOR_M3]
labels1  = [r'Config. $T_{1}$ ($M_{1}$)', r'Config. $T_{2} || T_{3}$ ($M_{3}$)']
markers1 = ['o', '^']

fig, ax = plt.subplots(figsize=(14, 10))
for data, ci, label, color, marker in zip([accuracy_m1, accuracy_m3_t1], 
                                        [ci_m1_f1, ci_m3_t1_f1],
                                        labels1, colors1, markers1):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
    ax.fill_between(rounds, 
                    [d - c for d, c in zip(data, ci)],
                    [d + c for d, c in zip(data, ci)],
                    color=color, alpha=0.2)
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('F1 Score', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper center', bbox_to_anchor=(0.5, 1.12), ncol=3)
plt.tight_layout()
plt.savefig('F1_Score.pdf', bbox_inches='tight')
plt.show()

# ------------------------------
# Grafico 2: MAE vs FL Round (m2 e m3_t2)
colors2  = [COLOR_M2, COLOR_M3]
labels2  = [r'Config. $T_{2}$ ($M_{2}$)', r'Config. $T_{2} || T_{3}$ ($M_{3}$)']
markers2 = ['s', 'd']

fig, ax = plt.subplots(figsize=(14, 10))
for data, ci, label, color, marker in zip([mae_m2, mae_m3_t2],
                                        [ci_m2_mae, ci_m3_t2_mae],
                                        labels2, colors2, markers2):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
    ax.fill_between(rounds,
                    [d - c for d, c in zip(data, ci)],
                    [d + c for d, c in zip(data, ci)],
                    color=color, alpha=0.2)
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('Mean Absolute Error (m)', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper center', bbox_to_anchor=(0.5, 1.12), ncol=2)
plt.tight_layout()
plt.savefig('MAE.pdf', bbox_inches='tight')
plt.show()

# ------------------------------
# Grafico 3: Average Training Time vs FL Round (m1 e m3)
labels_at1 = [r'Config. $T_{1}$ ($M_{1}$)', r'Config. $T_{2} || T_{3}$ ($M_{3}$)']
colors_at1 = [COLOR_M1, COLOR_M3]
markers_at1 = ['o', '^']

fig, ax = plt.subplots(figsize=(14, 10))
for data, ci, label, color, marker in zip([total_training_time_m1, total_training_time_m3_t1],
                                        [ci_m1_tt, ci_m3_tt],
                                        labels_at1, colors_at1, markers_at1):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
    ax.fill_between(rounds,
                    [d - c for d, c in zip(data, ci)],
                    [d + c for d, c in zip(data, ci)],
                    color=color, alpha=0.2)
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('Average Training Time (s)', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper center', bbox_to_anchor=(0.5, 1.12), ncol=2)
plt.tight_layout()
plt.savefig('TT_M1_M3.pdf', bbox_inches='tight')
plt.show()

# ------------------------------
# Grafico 4: Average Training Time vs FL Round (m2 e m3)
labels_at2 = [r'Config. $T_{2}$ ($M_{2}$)', r'Config. $T_{2} || T_{3}$ ($M_{3}$)']
colors_at2 = [COLOR_M2, COLOR_M3]
markers_at2 = ['s', 'd']

fig, ax = plt.subplots(figsize=(14, 10))
for data, ci, label, color, marker in zip([total_training_time_m2, total_training_time_m3_t1],
                                        [ci_m2_tt, ci_m3_tt],
                                        labels_at2, colors_at2, markers_at2):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
    ax.fill_between(rounds,
                    [d - c for d, c in zip(data, ci)],
                    [d + c for d, c in zip(data, ci)],
                    color=color, alpha=0.2)
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('Average Training Time (s)', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper center', bbox_to_anchor=(0.5, 1.12), ncol=2)
plt.tight_layout()
plt.savefig('TT_M2_M3.pdf', bbox_inches='tight')
plt.show()

# ------------------------------
# Grafico 5: F1 Score / Average Training Time (m1 e m3_t1)
fig, ax = plt.subplots(figsize=(14, 10))
for data, ci, label, color, marker in zip([f1_over_training_m1, f1_over_training_m3_t1],
                                        [ci_eff_m1, ci_eff_m3_t1],
                                        labels1, colors1, markers1):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
    ax.fill_between(rounds,
                    [d - c for d, c in zip(data, ci)],
                    [d + c for d, c in zip(data, ci)],
                    color=color, alpha=0.2)
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('F1 Score / Average Training Time', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper center', bbox_to_anchor=(0.5, 1.12), ncol=2)
plt.tight_layout()
plt.savefig('Efficiency_M1_M3.pdf', bbox_inches='tight')
plt.show()

# ------------------------------
# Grafico 6: MAE / Average Training Time (m2 e m3_t2)
fig, ax = plt.subplots(figsize=(14, 10))
for data, ci, label, color, marker in zip([mae_over_training_m2, mae_over_training_m3_t2],
                                        [ci_eff_m2, ci_eff_m3_t2],
                                        labels2, colors2, markers2):
    ax.plot(rounds, data, color=color, label=label, marker=marker, linewidth=2, linestyle='--')
    ax.fill_between(rounds,
                    [d - c for d, c in zip(data, ci)],
                    [d + c for d, c in zip(data, ci)],
                    color=color, alpha=0.2)
ax.set_xlabel('Federated Learning Round', fontsize=30)
ax.set_ylabel('MAE / Average Training Time', fontsize=30)
ax.tick_params(axis='both', labelsize=30)
ax.set_xticks(rounds)
ax.legend(fontsize=25, loc='upper center', bbox_to_anchor=(0.5, 1.12), ncol=2)
plt.tight_layout()
plt.savefig('Efficiency_M2_M3.pdf', bbox_inches='tight')
plt.show()
```
