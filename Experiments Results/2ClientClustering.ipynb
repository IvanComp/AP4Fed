```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager
import matplotlib.patches as mpatches  

# Define the rounds
rounds = np.arange(1, 11)
plt.rcParams['font.family'] = 'CMU Serif'

# Accuracy data provided
accuracy_4a4b_cluster = [0.12, 0.22, 0.29, 0.35, 0.38, 0.41, 0.44, 0.46, 0.47, 0.48]
accuracy_4a4b_no_cluster = [0.11, 0.12, 0.15, 0.17, 0.20, 0.19, 0.22, 0.21, 0.21, 0.22]
accuracy_5a3b_cluster = [0.08, 0.11, 0.23, 0.30, 0.35, 0.39, 0.40, 0.42, 0.44, 0.45]
accuracy_5a3b_no_cluster = [0.08, 0.05, 0.06, 0.13, 0.17, 0.16, 0.19, 0.17, 0.14, 0.19]
accuracy_6a2b_cluster = [0.06, 0.16, 0.23, 0.29, 0.34, 0.37, 0.40, 0.40, 0.44, 0.44]
accuracy_6a2b_no_cluster = [0.09, 0.09, 0.12, 0.14, 0.16, 0.17, 0.19, 0.19, 0.23, 0.21]

# Total Round Time data provided
total_round_time_4a4b_cluster = [240, 339, 269, 241, 306, 303, 254, 262, 236, 288]
total_round_time_4a4b_no_cluster = [451, 512, 504, 466, 433, 473, 455, 446, 477, 481]
total_round_time_5a3b_cluster = [371, 376, 298, 292, 369, 325, 344, 349, 338, 381]
total_round_time_5a3b_no_cluster = [213, 225, 216, 217, 208, 218, 242, 256, 235, 274]
total_round_time_6a2b_cluster = [393, 394, 344, 391, 398, 357, 308, 418, 364, 335]
total_round_time_6a2b_no_cluster = [391, 293, 305, 283, 327, 323, 330, 312, 318, 365]

# Calculate Accuracy over Total Round Time and multiply by 1000
accuracy_over_time_4a4b_cluster = [1000 * (a / t) for a, t in zip(accuracy_4a4b_cluster, total_round_time_4a4b_cluster)]
accuracy_over_time_4a4b_no_cluster = [1000 * (a / t) for a, t in zip(accuracy_4a4b_no_cluster, total_round_time_4a4b_no_cluster)]
accuracy_over_time_5a3b_cluster = [1000 * (a / t) for a, t in zip(accuracy_5a3b_cluster, total_round_time_5a3b_cluster)]
accuracy_over_time_5a3b_no_cluster = [1000 * (a / t) for a, t in zip(accuracy_5a3b_no_cluster, total_round_time_5a3b_no_cluster)]
accuracy_over_time_6a2b_cluster = [1000 * (a / t) for a, t in zip(accuracy_6a2b_cluster, total_round_time_6a2b_cluster)]
accuracy_over_time_6a2b_no_cluster = [1000 * (a / t) for a, t in zip(accuracy_6a2b_no_cluster, total_round_time_6a2b_no_cluster)]

# Colors and styles
color_4a4b_cluster = '#2ca02c'
color_4a4b_no_cluster = '#ff7f0e'
color_5a3b_cluster = '#1f77b4'
color_5a3b_no_cluster = '#d62728'
color_6a2b_cluster = '#9467bd'
color_6a2b_no_cluster = '#8c564b'

### First Plot: F1 Score for Model A ###
fig1, ax1 = plt.subplots(figsize=(10, 6))

# Area tratteggiata per "Without Cluster"
ax1.fill_between(rounds, 0.03, 0.26, color='lightcoral', alpha=0.15, label='Without Cluster', hatch='//')
ax1.fill_between(rounds, 0.4, 0.49, color='lightgreen', alpha=0.2, label='With Cluster')

# Patch per la legenda
hatch_patch1 = mpatches.Patch(facecolor='lightcoral', edgecolor='lightcoral', alpha=0.15, hatch='//', label='Without Cluster')
with_patch1 = mpatches.Patch(facecolor='lightgreen', edgecolor='lightgreen', alpha=0.2, label='With Cluster')

# Plot the curves
ax1.plot(rounds, accuracy_4a4b_cluster, color=color_4a4b_cluster, label='4a4b with cluster',
         marker='o', linewidth=2, linestyle='-')
ax1.plot(rounds, accuracy_4a4b_no_cluster, color=color_4a4b_no_cluster, label='4a4b no cluster',
         marker='s', linewidth=2, linestyle='--')
ax1.plot(rounds, accuracy_5a3b_cluster, color=color_5a3b_cluster, label='5a3b with cluster',
         marker='^', linewidth=2, linestyle=':')
ax1.plot(rounds, accuracy_5a3b_no_cluster, color=color_5a3b_no_cluster, label='5a3b no cluster',
         marker='d', linewidth=2, linestyle='-.')
ax1.plot(rounds, accuracy_6a2b_cluster, color=color_6a2b_cluster, label='6a2b with cluster',
         marker='*', linewidth=2, linestyle='-')
ax1.plot(rounds, accuracy_6a2b_no_cluster, color=color_6a2b_no_cluster, label='6a2b no cluster',
         marker='p', linewidth=2, linestyle='--')

# Labels, legend, and axis settings
ax1.set_xlabel('Federated Learning Round', fontsize=22)
ax1.set_ylabel('F1 Score for Model A', fontsize=22)
ax1.tick_params(axis='both', labelsize=22)
ax1.legend(handles=[hatch_patch1, with_patch1] + ax1.get_legend_handles_labels()[0][2:], fontsize=12, loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=4)
ax1.set_xticks(rounds)
ax1.set_ylim(0, 0.5)

plt.tight_layout()
plt.savefig('EnhancedClientClustering1.pdf', bbox_inches='tight', dpi=800)
plt.show()

### Second Plot: Total Round Time ###
fig2, ax2 = plt.subplots(figsize=(10, 6))

# Plot total round time
ax2.plot(rounds, total_round_time_4a4b_cluster, color=color_4a4b_cluster, label='4a4b with cluster',
         marker='o', linewidth=2, linestyle='-')
ax2.plot(rounds, total_round_time_4a4b_no_cluster, color=color_4a4b_no_cluster, label='4a4b no cluster',
         marker='s', linewidth=2, linestyle='--')
ax2.plot(rounds, total_round_time_5a3b_cluster, color=color_5a3b_cluster, label='5a3b with cluster',
         marker='^', linewidth=2, linestyle=':')
ax2.plot(rounds, total_round_time_5a3b_no_cluster, color=color_5a3b_no_cluster, label='5a3b no cluster',
         marker='d', linewidth=2, linestyle='-.')
ax2.plot(rounds, total_round_time_6a2b_cluster, color=color_6a2b_cluster, label='6a2b with cluster',
         marker='*', linewidth=2, linestyle='-')
ax2.plot(rounds, total_round_time_6a2b_no_cluster, color=color_6a2b_no_cluster, label='6a2b no cluster',
         marker='p', linewidth=2, linestyle='--')

# Labels, legend, and axis settings
ax2.set_xlabel('Federated Learning Round', fontsize=22)
ax2.set_ylabel('Total Round Time (s)', fontsize=22)
ax2.tick_params(axis='both', labelsize=22)
ax2.legend(fontsize=12, loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=3)
ax2.set_xticks(rounds)

plt.tight_layout()
plt.savefig('ClientClustering2.pdf', bbox_inches='tight', dpi=500)
plt.show()

fig3, ax3 = plt.subplots(figsize=(10, 6))

# Area tratteggiata per "Without Cluster"
ax3.fill_between(rounds, 0.1, 0.85, color='lightcoral', alpha=0.15, label='Without Cluster', hatch='//')
ax3.fill_between(rounds, 1.1, 2, color='lightgreen', alpha=0.2, label='With Cluster')

# Patch per la legenda
hatch_patch3 = mpatches.Patch(facecolor='lightcoral', edgecolor='lightcoral', alpha=0.15, hatch='//', label='Without Cluster')
with_patch3 = mpatches.Patch(facecolor='lightgreen', edgecolor='lightgreen', alpha=0.2, label='With Cluster')

# Plot accuracy over total round time, multiplied by 1000
ax3.plot(rounds, accuracy_over_time_4a4b_cluster, color=color_4a4b_cluster, label='4a4b with cluster',
         marker='o', linewidth=2, linestyle='-')
ax3.plot(rounds, accuracy_over_time_4a4b_no_cluster, color=color_4a4b_no_cluster, label='4a4b no cluster',
         marker='s', linewidth=2, linestyle='--')
ax3.plot(rounds, accuracy_over_time_5a3b_cluster, color=color_5a3b_cluster, label='5a3b with cluster',
         marker='^', linewidth=2, linestyle=':')
ax3.plot(rounds, accuracy_over_time_5a3b_no_cluster, color=color_5a3b_no_cluster, label='5a3b no cluster',
         marker='d', linewidth=2, linestyle='-.')
ax3.plot(rounds, accuracy_over_time_6a2b_cluster, color=color_6a2b_cluster, label='6a2b with cluster',
         marker='*', linewidth=2, linestyle='-')
ax3.plot(rounds, accuracy_over_time_6a2b_no_cluster, color=color_6a2b_no_cluster, label='6a2b no cluster',
         marker='p', linewidth=2, linestyle='--')

# Labels, legend, and axis settings
ax3.set_xlabel('Federated Learning Round', fontsize=22)
ax3.set_ylabel('Accuracy / Total Round Time', fontsize=22)
ax3.tick_params(axis='both', labelsize=22)
ax3.set_xticks(rounds)

# Legenda completa con i patch per le aree tratteggiate
ax3.legend(handles=[hatch_patch3, with_patch3] + ax3.get_legend_handles_labels()[0][2:], fontsize=12, loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=4)

plt.tight_layout()
plt.savefig('ClientClustering3.svg', bbox_inches='tight', dpi=300)
plt.show()
```
