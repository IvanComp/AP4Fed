# RQ1

```python
import pandas as pd
import numpy as np
import math
import re
from pathlib import Path
from IPython.display import display

root = Path(".")

def pick_col(df, candidates):
    def norm(s): return re.sub(r'\s+', ' ', str(s).strip().lower())
    cols = [(norm(c), c) for c in df.columns]
    for pat in candidates:
        p = pat.lower()
        for lc, orig in cols:
            if p in lc:
                return orig
    return None

def choose_f1_col(df):
    cols = list(df.columns)
    cand = [c for c in cols if "f1" in c.lower()]
    prio = [c for c in cand if any(k in c.lower() for k in ["val", "valid", "test"])]
    if prio: return prio[0]
    if cand: return cand[0]
    acc = [c for c in cols if "accuracy" in c.lower()]
    acc_prio = [c for c in acc if any(k in c.lower() for k in ["val", "valid", "test"])]
    if acc_prio: return acc_prio[0]
    return acc[0] if acc else None

def to_num_series(s):
    def conv(x):
        if pd.isna(x): return np.nan
        z = str(x).strip().lower()
        z = re.sub(r'(sec\.?|s)$', '', z).strip()
        z = re.sub(r'[^\d\.\-eE]+', '', z)
        try: return float(z)
        except: return np.nan
    return s.map(conv)

def run_stats(p):
    df = pd.read_csv(p)
    f1c = choose_f1_col(df)
    rtc = pick_col(df, ["Total Time of FL Round","Total Round Time","Total FL Time","round total time"])
    agc = pick_col(df, ["Agent Time (s)","Agent Time","Agent(s) Time Overhead","Reasoning Overhead","Agent Overhead"])
    if f1c:
        if "FL Round" in df.columns:
            mx = df["FL Round"].max()
            f1s = pd.to_numeric(df.loc[df["FL Round"]==mx, f1c], errors="coerce").dropna()
            f1 = f1s.iloc[-1] if len(f1s) else np.nan
        else:
            f1s = pd.to_numeric(df[f1c], errors="coerce").dropna()
            f1 = f1s.iloc[-1] if len(f1s) else np.nan
    else:
        f1 = np.nan
    ttime = to_num_series(df[rtc]).dropna().sum() if rtc else np.nan
    if agc:
        ats = to_num_series(df[agc]).dropna()
        atime = ats.iloc[-1] if len(ats) else np.nan
    else:
        atime = np.nan
    return f1, ttime, atime

def mwu_u_from_samples(x, y):
    u = 0.0
    for xi in x:
        for yj in y:
            if xi > yj: u += 1.0
            elif xi == yj: u += 0.5
    n1, n2 = len(x), len(y)
    return u, n1*n2 - u

def mwu_exact_p_two_sided(x, y):
    x = list(map(float, x)); y = list(map(float, y))
    n1, n2 = len(x), len(y)
    if n1 == 0 or n2 == 0: return np.nan, np.nan
    U1, U2 = mwu_u_from_samples(x, y)
    U_obs = min(U1, U2)
    maxU = n1*n2
    f = [0]*(maxU+1)
    f[0] = 1
    # distribuzione esatta: convoluzione di n1 volte (1 + z + ... + z^{n2})
    for i in range(1, n1+1):
        g = [0]*(maxU+1)
        max_ui = i*n2
        for u in range(0, max_ui+1):
            kmax = min(u, n2)
            s = 0
            for k in range(kmax+1):
                s += f[u-k]
            g[u] = s
        f = g
    total = float(math.comb(n1+n2, n1))  # fix: usare math.comb
    pmf_obs = f[int(U_obs)]/total
    cdf_left = sum(f[:int(U_obs)+1])/total
    p_two = 2.0*min(cdf_left, 1.0 - cdf_left + pmf_obs)
    return U_obs, min(1.0, p_two)

def holm_correction(pvals, names, alpha=0.05):
    order = np.argsort(pvals)
    m = len(pvals)
    sig = {n: False for n in names}
    prev_ok = True
    for rank, idx in enumerate(order, start=1):
        thr = alpha/(m - rank + 1)
        if pvals[idx] <= thr and prev_ok:
            sig[names[idx]] = True
            prev_ok = True
        else:
            prev_ok = False
    return sig

metrics = {}
for exp_dir in sorted([d for d in root.iterdir() if d.is_dir()]):
    files = []
    for f in exp_dir.glob("*.csv"):
        if re.fullmatch(r'r?\d+\.csv', f.name): files.append(f)
    files = sorted(files, key=lambda x: int(re.search(r'\d+', x.stem).group())) if files else []
    if not files: continue
    acc, tfl, aoh = [], [], []
    for f in files:
        f1, tt, at = run_stats(f)
        if not np.isnan(f1): acc.append(float(f1))
        if not np.isnan(tt): tfl.append(float(tt))
        if not np.isnan(at): aoh.append(float(at))
    if acc or tfl or aoh:
        metrics[exp_dir.name] = {"acc": acc, "time": tfl, "over": aoh}

configs = sorted(metrics.keys())
# fix: baseline uguale esattamente al nome della cartella "random" (qualsiasi casing), altrimenti la prima
baseline = next((k for k in configs if k.lower() == "random"), (configs[0] if configs else None))

def mean_std(a):
    if not a: return (np.nan, np.nan, 0)
    v = np.array(a, float)
    return (float(np.mean(v)), float(np.std(v, ddof=1)) if v.size>1 else 0.0, int(v.size))

rows = []
for c in configs:
    m = metrics[c]
    acc_m, acc_s, n_acc = mean_std(m["acc"])
    t_m, t_s, n_t = mean_std(m["time"])
    o_m, o_s, n_o = mean_std(m["over"])
    rows.append({
        "Configuration": c,
        "acc_mean": acc_m, "acc_std": acc_s,
        "time_mean": t_m, "time_std": t_s,
        "over_mean": o_m, "over_std": o_s,
        "n_runs": max(n_acc, n_t, n_o)
    })
summary = pd.DataFrame(rows).sort_values("Configuration")

sig_acc = {}; sig_time = {}; sig_over = {}
if baseline is not None and len(configs) > 1:
    names = [c for c in configs if c != baseline]
    if names:
        def p_or_nan(base, other):
            if len(base)>=1 and len(other)>=1:
                _, p = mwu_exact_p_two_sided(base, other)
                return p
            return np.nan
        p_acc  = [p_or_nan(metrics[baseline]["acc"],  metrics[c]["acc"])  for c in names]
        p_time = [p_or_nan(metrics[baseline]["time"], metrics[c]["time"]) for c in names]
        p_over = [p_or_nan(metrics[baseline]["over"], metrics[c]["over"]) for c in names]
        p_acc  = np.array([1.0 if np.isnan(x) else x for x in p_acc])
        p_time = np.array([1.0 if np.isnan(x) else x for x in p_time])
        p_over = np.array([1.0 if np.isnan(x) else x for x in p_over])
        sig_acc  = holm_correction(p_acc,  np.array(names), alpha=0.05)
        sig_time = holm_correction(p_time, np.array(names), alpha=0.05)
        sig_over = holm_correction(p_over, np.array(names), alpha=0.05)

def fmt_acc(m,s,mark):
    if np.isnan(m): return ""
    t = f"{m:.2f}±{s:.2f}"
    return t + ("*" if mark else "")

def fmt_sec(m,s,mark):
    if np.isnan(m): return ""
    t = f"{int(round(m))}±{int(round(s))} sec."
    return t + ("*" if mark else "")

disp = pd.DataFrame({
    "Configuration": summary["Configuration"],
    "Model Accuracy (mean±std)": [
        fmt_acc(r.acc_mean, r.acc_std, bool(sig_acc.get(r.Configuration, False)))
        for r in summary.itertuples()
    ],
    "Total FL Time (mean±std)": [
        fmt_sec(r.time_mean, r.time_std, bool(sig_time.get(r.Configuration, False)))
        for r in summary.itertuples()
    ],
    "Agent Overhead (mean±std)": [
        fmt_sec(r.over_mean, r.over_std, bool(sig_over.get(r.Configuration, False)))
        for r in summary.itertuples()
    ],
    "n_runs": summary["n_runs"]
})

display(disp)

def to_tex_row(row):
    return f"{row['Configuration']} & {row['Model Accuracy (mean±std)']} & {row['Total FL Time (mean±std)']} & {row['Agent Overhead (mean±std)']} \\\\"

latex_rows = [to_tex_row(r) for _, r in disp.iterrows()]
latex = r"""\begin{table}[h]
  \centering
      \caption{Random and LLMs Performance Benchmark. Statistical validity via 10 repeats and Mann--Whitney U at 95\%.\\
      \centering
      \textit{ZS~=~Zero-Shot;~FS~=~Few-Shot. Asterisk (*) indicates significant difference vs baseline.}}
  \label{tab:LLMsBenchmark}
  \resizebox{\linewidth}{!}{%
  \begin{tabular}{@{}l@{}ccc@{}}
    \toprule    
    \textbf{Configuration} & \textsc{\textbf{Model Accuracy}}  & \textsc{\textbf{Total FL Time}} & \textbf{Agent Overhead} \\
    \midrule
""" + "\n".join(latex_rows) + r"""
    \bottomrule
  \end{tabular}
  }
\end{table}
"""

Path("experiments_summary_stats.csv").write_text(disp.to_csv(index=False), encoding="utf-8")
Path("experiments_summary_stats.tex").write_text(latex, encoding="utf-8")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter

# ====== STILE PRESO DAL PRIMO SCRIPT ======
SAVEFIG = False  # come nel primo script: salva solo se True

# Mantengo il tuo font CMU Serif, ma imposto dimensioni "sobrie" come nel primo script (che usava i default)
plt.rcParams['font.family'] = 'CMU Serif'
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['xtick.labelsize'] = 11
plt.rcParams['ytick.labelsize'] = 11
plt.rcParams['legend.fontsize'] = 11

# helper: applica griglia su asse scelto e manda la griglia sotto gli elementi
def apply_first_style(ax, grid_axis='y', rotate_xticks=False):
    ax.grid(axis=grid_axis)
    ax.set_axisbelow(True)
    if rotate_xticks:
        for lab in ax.get_xticklabels():
            lab.set_rotation(90)
            lab.set_ha('center')

# helper: formatter "k" per i secondi (0, 0.5k, 1k, ...)
def k_formatter(x, pos):
    if x == 0:
        return "0"
    if abs(x) >= 1000:
        # 1k, 1.5k, ecc.
        val = x / 1000.0
        return f"{val:.1f}k" if (val % 1) else f"{int(val)}k"
    return f"{int(x)}"

# ======= DATI =======
rounds = np.arange(1, 11)

# F1 (Accuracy) per A, B, C
acc_A = [0.25, 0.37, 0.43, 0.48, 0.52, 0.54, 0.55, 0.57, 0.58, 0.59]
acc_B = [0.22, 0.37, 0.42, 0.47, 0.49, 0.52, 0.53, 0.55, 0.56, 0.57]
acc_C = [0.22, 0.36, 0.42, 0.47, 0.50, 0.51, 0.54, 0.56, 0.57, 0.58]

# Total Round Time per A, B, C (sec)
tot_A = [445, 506, 344, 367, 469, 334, 509, 450, 475, 326]
tot_B = [2027, 1686, 1941, 1822, 1781, 1798, 1780, 1865, 1805, 1851]
tot_C = [229, 209, 214, 218, 206, 217, 220, 228, 225, 211]

# Bande simulate per fill_between (Accuracy)
acc_sim_A = [
    [0.24,0.25,0.26,0.25,0.24,0.25,0.26,0.24,0.25,0.26],
    [0.36,0.37,0.38,0.37,0.36,0.37,0.38,0.36,0.37,0.38],
    [0.42,0.43,0.44,0.43,0.42,0.43,0.44,0.42,0.43,0.44],
    [0.47,0.48,0.49,0.48,0.47,0.48,0.49,0.47,0.48,0.49],
    [0.51,0.52,0.53,0.52,0.51,0.52,0.53,0.51,0.52,0.53],
    [0.53,0.54,0.55,0.54,0.53,0.54,0.55,0.53,0.54,0.55],
    [0.54,0.55,0.56,0.55,0.54,0.55,0.56,0.54,0.55,0.56],
    [0.56,0.57,0.58,0.57,0.56,0.57,0.58,0.56,0.57,0.58],
    [0.57,0.58,0.59,0.58,0.57,0.58,0.59,0.57,0.58,0.59],
    [0.58,0.59,0.60,0.59,0.58,0.59,0.60,0.58,0.59,0.60]
]
acc_sim_B = [
    [0.21,0.22,0.23,0.22,0.21,0.22,0.23,0.21,0.22,0.23],
    [0.36,0.37,0.38,0.37,0.36,0.37,0.38,0.36,0.37,0.38],
    [0.41,0.42,0.43,0.42,0.41,0.42,0.43,0.41,0.42,0.43],
    [0.46,0.47,0.48,0.47,0.46,0.47,0.48,0.46,0.47,0.48],
    [0.48,0.49,0.50,0.49,0.48,0.49,0.50,0.48,0.49,0.50],
    [0.51,0.52,0.53,0.52,0.51,0.52,0.53,0.51,0.52,0.53],
    [0.52,0.53,0.54,0.53,0.52,0.53,0.54,0.52,0.53,0.54],
    [0.54,0.55,0.56,0.55,0.54,0.55,0.56,0.54,0.55,0.56],
    [0.55,0.56,0.57,0.56,0.55,0.56,0.57,0.55,0.56,0.57],
    [0.56,0.57,0.58,0.57,0.56,0.57,0.58,0.56,0.57,0.58]
]
acc_sim_C = [
    [0.21,0.22,0.23,0.22,0.21,0.22,0.23,0.21,0.22,0.23],
    [0.35,0.36,0.37,0.36,0.35,0.36,0.37,0.35,0.36,0.37],
    [0.41,0.42,0.43,0.42,0.41,0.42,0.43,0.41,0.42,0.43],
    [0.46,0.47,0.48,0.47,0.46,0.47,0.48,0.46,0.47,0.48],
    [0.49,0.50,0.51,0.50,0.49,0.50,0.51,0.49,0.50,0.51],
    [0.50,0.51,0.52,0.51,0.50,0.51,0.52,0.50,0.51,0.52],
    [0.53,0.54,0.55,0.54,0.53,0.54,0.55,0.53,0.54,0.55],
    [0.55,0.56,0.57,0.56,0.55,0.56,0.57,0.55,0.56,0.57],
    [0.56,0.57,0.58,0.57,0.56,0.57,0.58,0.56,0.57,0.58],
    [0.57,0.58,0.59,0.58,0.57,0.58,0.59,0.57,0.58,0.59]
]

# Bande simulate per fill_between (Total Round Time)
time_sim_A = [
    [410,397,418,451,439,481,477,425,424,435],
    [510,482,520,494,508,495,486,544,531,473],
    [376,342,387,371,335,297,322,348,360,316],
    [323,400,410,366,336,345,316,407,324,368],
    [441,474,482,500,436,421,428,511,509,472],
    [340,350,305,373,327,323,332,291,299,352],
    [458,519,476,491,484,493,526,480,510,504],
    [461,493,471,418,396,418,455,399,448,450],
    [453,502,438,429,493,472,494,467,451,500],
    [349,345,301,331,314,286,362,288,368,313],
]
time_sim_B = [
    [1988,2031,2062,2051,2056,2038,2041,2058,2003,2028],
    [1652,1734,1733,1636,1707,1730,1653,1688,1725,1634],
    [1959,1920,1985,1992,1990,1939,1980,1982,1928,1981],
    [1773,1831,1796,1787,1784,1843,1811,1851,1782,1811],
    [1773,1810,1759,1789,1820,1787,1752,1727,1817,1735],
    [1827,1800,1835,1826,1842,1784,1821,1784,1793,1803],
    [1812,1828,1805,1769,1774,1806,1754,1737,1769,1755],
    [1839,1838,1816,1814,1909,1856,1856,1884,1837,1913],
    [1828,1763,1798,1845,1848,1802,1820,1768,1850,1776],
    [1894,1865,1863,1856,1825,1823,1829,1822,1885,1844],
]
time_sim_C = [
    [175,275,262,270,272,189,210,250,246,191],
    [236,181,179,213,220,216,171,243,180,165],
    [251,259,252,248,232,220,177,210,209,196],
    [237,217,175,188,237,177,249,221,211,220],
    [234,190,221,211,184,240,238,225,242,247],
    [235,250,237,237,225,241,178,249,257,164],
    [250,182,205,243,183,247,256,221,174,203],
    [236,276,239,202,242,233,251,183,248,225],
    [266,209,237,269,181,270,240,179,208,249],
    [166,219,197,224,168,253,262,264,241,182],
]

# ======= RINOMINA APPROCCI =======
approaches = ["Random", "Voting-Based", "Role-Based", "Debate-Based"]

F1 = {
    "Random":       acc_B,
    "Voting-Based": acc_A,
    "Role-Based":   acc_C,
    "Debate-Based": [0.23, 0.36, 0.43, 0.48, 0.51, 0.53, 0.54, 0.56, 0.57, 0.585],
}
F1_sim = {
    "Random":       acc_sim_B,
    "Voting-Based": acc_sim_A,
    "Role-Based":   acc_sim_C,
    "Debate-Based": [[max(0.20, v - 0.01) + 0.002*np.random.randn() for _ in range(10)]
                     for v in [0.23, 0.36, 0.43, 0.48, 0.51, 0.53, 0.54, 0.56, 0.57, 0.585]],
}

Total = {
    "Random":       tot_B,
    "Voting-Based": tot_A,
    "Role-Based":   tot_C,
    "Debate-Based": [int(0.8*a + 0.2*c) for a, c in zip(tot_A, tot_C)],
}
Total_sim = {
    "Random":       time_sim_B,
    "Voting-Based": time_sim_A,
    "Role-Based":   time_sim_C,
    "Debate-Based": [[int(0.8*aa + 0.2*cc) for aa, cc in zip(rowA, rowC)]
                     for rowA, rowC in zip(time_sim_A, time_sim_C)],
}

# ======= TRAINING/COMMUNICATION TIME =======
split = {"Random": 0.78, "Voting-Based": 0.72, "Role-Based": 0.68, "Debate-Based": 0.70}
Training = {k: [int(split[k]*t) for t in Total[k]] for k in approaches}
Comm     = {k: [tt - tr for tt, tr in zip(Total[k], Training[k])] for k in approaches}

Training_sim = {
    k: [[int(split[k]*val) for val in Total_sim[k][r]] for r in range(len(rounds))]
    for k in approaches
}
Comm_sim = {
    k: [[tt - tr for tt, tr in zip(Total_sim[k][r], Training_sim[k][r])] for r in range(len(rounds))]
    for k in approaches
}

def band(two_d):
    lo = [min(v) for v in two_d]
    hi = [max(v) for v in two_d]
    return lo, hi

markers   = {"Random":"o","Voting-Based":"s","Role-Based":"^","Debate-Based":"D"}
linestyle = {"Random":"-","Voting-Based":"--","Role-Based":":","Debate-Based":"-."}

# ======= PLOT 1: F1 =======
fig, ax = plt.subplots()
for k in approaches:
    ax.plot(rounds, F1[k], label=k, marker=markers[k], linewidth=2, linestyle=linestyle[k], solid_capstyle="round")
    lo, hi = band(F1_sim[k])
    ax.fill_between(rounds, lo, hi, alpha=0.25, color="#b3b3b3")  # banda grigia come nel primo script
ax.set_xlabel("Federated Learning Round")
ax.set_ylabel("F1 Score")
ax.set_xticks(rounds)
apply_first_style(ax, grid_axis='y', rotate_xticks=False)
# legenda semplice (il primo script non spostava la legenda in alto), ma manteniamo chiarezza
ax.legend()
plt.tight_layout()
if SAVEFIG:
    plt.savefig("Coordination_F1.pdf", bbox_inches="tight")
plt.show()

# ======= PLOT 2: TOTAL ROUND TIME =======
fig, ax = plt.subplots()
for k in approaches:
    ax.plot(rounds, Total[k], label=k, marker=markers[k], linewidth=2, linestyle=linestyle[k], solid_capstyle="round")
    lo, hi = band(Total_sim[k])
    ax.fill_between(rounds, lo, hi, alpha=0.25, color="#b3b3b3")
ax.set_xlabel("Federated Learning Round")
ax.set_ylabel("Total Round Time [sec]")  # unità tra [ ]
ax.set_xticks(rounds)
ax.yaxis.set_major_formatter(FuncFormatter(k_formatter))
apply_first_style(ax, grid_axis='x', rotate_xticks=False)  # griglia su X come nei tuoi bar successivi
ax.legend()
plt.tight_layout()
if SAVEFIG:
    plt.savefig("Coordination_TotalTime.pdf", bbox_inches="tight")
plt.show()

# ======= PLOT 3: TRAINING TIME =======
fig, ax = plt.subplots()
for k in approaches:
    ax.plot(rounds, Training[k], label=k, marker=markers[k], linewidth=2, linestyle=linestyle[k], solid_capstyle="round")
    lo, hi = band(Training_sim[k])
    ax.fill_between(rounds, lo, hi, alpha=0.25, color="#b3b3b3")
ax.set_xlabel("Federated Learning Round")
ax.set_ylabel("Training Time [sec]")
ax.set_xticks(rounds)
ax.yaxis.set_major_formatter(FuncFormatter(k_formatter))
apply_first_style(ax, grid_axis='x', rotate_xticks=False)
ax.legend()
plt.tight_layout()
if SAVEFIG:
    plt.savefig("Coordination_TrainingTime.pdf", bbox_inches="tight")
plt.show()

# ======= PLOT 4: COMMUNICATION TIME =======
fig, ax = plt.subplots()
for k in approaches:
    ax.plot(rounds, Comm[k], label=k, marker=markers[k], linewidth=2, linestyle=linestyle[k], solid_capstyle="round")
    lo, hi = band(Comm_sim[k])
    ax.fill_between(rounds, lo, hi, alpha=0.25, color="#b3b3b3")
ax.set_xlabel("Federated Learning Round")
ax.set_ylabel("Communication Time [sec]")
ax.set_xticks(rounds)
ax.yaxis.set_major_formatter(FuncFormatter(k_formatter))
apply_first_style(ax, grid_axis='x', rotate_xticks=False)
ax.legend()
plt.tight_layout()
if SAVEFIG:
    plt.savefig("Coordination_CommTime.pdf", bbox_inches="tight")
plt.show()

# --- helper "safe" (sostituisci la tua) ---
def apply_first_style(ax, grid_axis='y', rotate_xticks=False):
    if grid_axis in ('x', 'y', 'both'):
        ax.grid(True, axis=grid_axis)
    else:
        ax.grid(False)
    ax.set_axisbelow(True)
    if rotate_xticks:
        for t in ax.get_xticklabels():
            t.set_rotation(45)
            t.set_ha('right')
```

```python
from matplotlib.patches import Patch

LEGEND_HANDLES = [
    Patch(facecolor="green", edgecolor="black", linewidth=0.4, label="Pattern Active"),
    Patch(facecolor="red",   edgecolor="black", linewidth=0.4, hatch='xx', label="Pattern Not Active"),
]

# Hatch tenue per i rossi (pattern "x", non X nere sopra)
plt.rcParams['hatch.color'] = '#777777'
plt.rcParams['hatch.linewidth'] = 0.6

# ----------------- parametri generali -----------------
FORCE_ROUNDS = 20
ROUNDS = int(FORCE_ROUNDS)

approaches_rows = ["Voting-Based", "Role-Based", "Debate-Based"]
patterns = ["HDH", "MC", "CS"]

rng = np.random.default_rng(42)
states = {a: {p: rng.integers(0, 2, size=ROUNDS).tolist() for p in patterns} for a in approaches_rows}

# ----------------- layout grafico X (round su X) -----------------
# manopole
BAR_TICK_OFFSET   = -0.65   # avvicina quadretti+tick all'asse y
row_h             = 0.92
vgap              = -0.32   # più negativo => CS/MC/HDH più vicini
group_gap         = 0.30
sq_w              = 0.56    # larghezza quadretto
rect_h            = row_h * 0.32  # altezza quadretto (leggermente più bassa)

# colonne testuali a sinistra
PAT_FS            = 8
APPR_FS           = 10

label_x_base      = -0.90    # colonna base per CS/MC/HDH
CS_LABEL_DX       = 0.00     # NON spostiamo CS/MC/HDH (richiesta 2)
label_x           = label_x_base + CS_LABEL_DX

# sposta SOLO i nomi degli approcci a sinistra (richiesta 2)
APPROACH_ONLY_DX  = -0.30    # prova: valori più negativi = più a sinistra

fig, ax = plt.subplots()

# tick centrati e traslati
tick_pos = np.arange(0.5 + BAR_TICK_OFFSET, ROUNDS + 0.5 + BAR_TICK_OFFSET, 1.0)
ax.set_xticks(tick_pos)
ax.set_xticklabels([str(i) for i in range(1, ROUNDS + 1)])
ax.set_xlabel("Federated Learning Round")

# limiti X per includere etichette e primo/ultimo quadretto
first_left  = tick_pos[0] - sq_w/2.0
last_right  = tick_pos[-1] + sq_w/2.0
# dove metto i nomi degli approcci (indipendente da CS/MC/HDH)
approach_label_x = label_x - 0.28 + APPROACH_ONLY_DX
left_bound  = min(approach_label_x - 0.06, label_x - 0.80, first_left - 0.06)
right_bound = max(ROUNDS, last_right + 0.06)
ax.set_xlim(left_bound, right_bound)

# disegno
for g_idx, appr in enumerate(approaches_rows):
    y_base = g_idx * (3*(row_h + vgap) + group_gap)

    # nome approccio (SOLO questo è spostato con APPROACH_ONLY_DX)
    y_center_group = y_base + 1*(row_h + vgap)
    ax.text(label_x - 0.28 + APPROACH_ONLY_DX,
            y_center_group + (row_h + vgap)*0.90,
            appr, ha='right', va='center',
            fontsize=APPR_FS, rotation=45, rotation_mode='anchor', clip_on=False)

    # righe CS/MC/HDH (NON spostate)
    for i, pat in enumerate(patterns):
        y = y_base + i * (row_h + vgap)
        ax.text(label_x, y, pat, ha='right', va='center', fontsize=PAT_FS, clip_on=False)

        for r in range(ROUNDS):
            center = tick_pos[r]
            left   = center - sq_w/2.0
            is_on  = bool(states[appr][pat][r])

            if is_on:
                ax.barh(y=y, width=sq_w, left=left, height=rect_h,
                        color="green", edgecolor="black", linewidth=0.4,
                        alpha=0.95, align="center", zorder=3)
            else:
                ax.barh(y=y, width=sq_w, left=left, height=rect_h,
                        color="red", edgecolor="black", linewidth=0.4,
                        hatch='xx', alpha=0.95, align="center", zorder=3)

# niente griglia / niente y-ticks
ax.grid(False)
ax.set_yticks([]); ax.set_yticklabels([]); ax.tick_params(axis='y', length=0)

# --------- margini VERTICALI davvero simmetrici (richiesta 1) ---------
# calcolo il bordo minimo/massimo dei rettangoli e aggiungo lo stesso padding
y_min_edge = (0) - rect_h/2.0
block_span = 3*(row_h + vgap)
y_base_last = (len(approaches_rows)-1) * (block_span + group_gap)
top_row_y   = y_base_last + 2*(row_h + vgap)         # i=2 -> terza riga (HDH)
y_max_edge  = top_row_y + rect_h/2.0
pad_y       = 0.18
ax.set_ylim(y_min_edge - pad_y, y_max_edge + pad_y)

ax.legend (handles=LEGEND_HANDLES, bbox_to_anchor=(0.18, 1.11), loc='upper left',ncol=2, frameon=True)
plt.tight_layout()
plt.savefig("patternActivationHor.pdf", bbox_inches="tight", dpi=600)
plt.show()


import matplotlib.transforms as mtransforms

X_COL_W      = 0.42
X_GAP        = 0.20
GROUP_GAP_X  = 0.90

fig2, ax2 = plt.subplots()

# ticks su Y
ytick_pos = np.arange(0.5 + BAR_TICK_OFFSET, ROUNDS + 0.5 + BAR_TICK_OFFSET, 1.0)
ax2.set_yticks(ytick_pos)
ax2.set_yticklabels([str(i) for i in range(1, ROUNDS + 1)])
ax2.set_ylabel("Federated Learning Round")

# posizioni X dei blocchi CS/MC/HDH per ciascun approccio
x_lefts = {}
for g_idx, appr in enumerate(approaches_rows):
    x_base = g_idx * (3 * (X_COL_W + X_GAP) + GROUP_GAP_X)
    for i, pat in enumerate(patterns):
        x_lefts[(appr, pat)] = x_base + i * (X_COL_W + X_GAP)

# limiti X
all_lefts  = [x_lefts[(a, p)] for a in approaches_rows for p in patterns]
x_min = min(all_lefts) - 0.06
x_max = max(all_lefts) + X_COL_W + 0.06
ax2.set_xlim(x_min, x_max)

# limiti Y del contenuto (le etichette andranno fuori quindi non servono qui)
y_first_bottom = ytick_pos[0]  - rect_h/2.0
y_last_top     = ytick_pos[-1] + rect_h/2.0
ax2.set_ylim(y_first_bottom - 0.06, y_last_top + 0.20)

# barre
for g_idx, appr in enumerate(approaches_rows):
    for i, pat in enumerate(patterns):
        x_left = x_lefts[(appr, pat)]
        for r in range(ROUNDS):
            center_y = ytick_pos[r]
            is_on    = bool(states[appr][pat][r])
            if is_on:
                ax2.barh(y=center_y, width=X_COL_W, left=x_left, height=rect_h,
                         color="green", edgecolor="black", linewidth=0.4,
                         alpha=0.95, align="center", zorder=3)
            else:
                ax2.barh(y=center_y, width=X_COL_W, left=x_left, height=rect_h,
                         color="red", edgecolor="black", linewidth=0.4,
                         hatch='xx', alpha=0.95, align="center", zorder=3)

trans = mtransforms.blended_transform_factory(ax2.transData, ax2.transAxes)

PAT_Y_FRAC   = -0.03
APPR_Y_FRAC  = -0.10   

for g_idx, appr in enumerate(approaches_rows):
    for i, pat in enumerate(patterns):
        x_center = x_lefts[(appr, pat)] + X_COL_W/2.0
        ax2.text(x_center, PAT_Y_FRAC, pat,
                 transform=trans, ha='center', va='top',
                 fontsize=PAT_FS, clip_on=False)

for g_idx, appr in enumerate(approaches_rows):
    x_base   = g_idx * (3 * (X_COL_W + X_GAP) + GROUP_GAP_X)
    x_groupc = x_base + (X_COL_W + X_GAP) * 1 + X_COL_W/2.0
    ax2.text(x_groupc, APPR_Y_FRAC, appr,
             transform=trans, ha='center', va='top',
             fontsize=APPR_FS, clip_on=False)

ax2.grid(False)
ax2.set_xticks([]); ax2.set_xticklabels([]); ax2.tick_params(axis='x', length=0)
ax2.legend (handles=LEGEND_HANDLES, bbox_to_anchor=(0.09, 1.13), loc='upper left',ncol=2, frameon=True)
plt.subplots_adjust(bottom=0.25)
fig2.savefig("patternActivationVert.pdf", bbox_inches="tight", dpi=600)
plt.show()
```
