```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'CMU Serif'

rounds = np.arange(1, 11)  # 10 round

accuracy_P1 = [0.25, 0.37, 0.43, 0.48, 0.52, 0.54, 0.55, 0.57, 0.58, 0.59]
accuracy_P1_prime = [0.22, 0.37, 0.42, 0.47, 0.49, 0.52, 0.53, 0.55, 0.56, 0.57]
accuracy_P1_excluded = [0.22, 0.36, 0.42, 0.47, 0.50, 0.51, 0.54, 0.56, 0.57, 0.58]

time_P1 = [445, 506, 344, 367, 469, 334, 509, 450, 475, 326]
time_P1_prime = [2027, 1686, 1941, 1822, 1781, 1798, 1780, 1865, 1805, 1851]
time_P1_excluded = [229, 209, 214, 218, 206, 217, 220, 228, 225, 211]

efficiency_P1 = [0.06, 0.07, 0.13, 0.13, 0.11, 0.16, 0.11, 0.13, 0.12, 0.18]
efficiency_P1_prime = [0.01, 0.02, 0.02, 0.03, 0.03, 0.03, 0.03, 0.03, 0.03, 0.03]
efficiency_P1_excluded = [0.10, 0.17, 0.20, 0.22, 0.24, 0.24, 0.25, 0.25, 0.25, 0.27]

color_p1 = '#1f77b4'  
color_p1_prime = '#d62728'  
color_p1_excluded = '#2ca02c' 

# Accuracy x Rounds
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.plot(rounds, accuracy_P1, color=color_p1, label='Config. A', marker='o', linewidth=2, solid_capstyle='round', linestyle='-')
ax1.plot(rounds, accuracy_P1_prime, color=color_p1_prime, label='Config. B', marker='s', linewidth=2, solid_capstyle='round', linestyle='--')
ax1.plot(rounds, accuracy_P1_excluded, color=color_p1_excluded, label='Config. C', marker='^', linewidth=2, solid_capstyle='round', linestyle=':')

ax1.set_xlabel('Federated Learning Round', fontsize=20)
ax1.set_ylabel('F1 Score', fontsize=20)
ax1.legend(fontsize=25)
ax1.set_xticks(rounds)  
ax1.tick_params(axis='both', labelsize=20)
plt.tight_layout()
plt.savefig('ClientSelector1.pdf')
plt.show()

# Total Round Time x Rounds
fig, ax2 = plt.subplots(figsize=(10, 6))
ax2.plot(rounds, time_P1, color=color_p1, label='Config. A', marker='o', linewidth=2, solid_capstyle='round', linestyle='-')
ax2.plot(rounds, time_P1_prime, color=color_p1_prime, label='Config. B', marker='s', linewidth=2, solid_capstyle='round', linestyle='--')
ax2.plot(rounds, time_P1_excluded, color=color_p1_excluded, label='Config. C', marker='^', linewidth=2, solid_capstyle='round', linestyle=':')

ax2.set_xlabel('Federated Learning Round', fontsize=20)
ax2.set_ylabel('Total Round Time (s)', fontsize=20)
#ax2.legend(fontsize=20)
ax2.set_xticks(rounds)  
ax2.tick_params(axis='both', labelsize=20)
plt.tight_layout()
plt.savefig('ClientSelector2.pdf')
plt.show()

# Accuracy over Efficiency
fig, ax3 = plt.subplots(figsize=(10, 6))
ax3.plot(rounds, efficiency_P1, color=color_p1, label='Config. A', marker='o', linewidth=2, solid_capstyle='round', linestyle='-')
ax3.plot(rounds, efficiency_P1_prime, color=color_p1_prime, label='Config. B', marker='s', linewidth=2, solid_capstyle='round', linestyle='--')
ax3.plot(rounds, efficiency_P1_excluded, color=color_p1_excluded, label='Config. C', marker='^', linewidth=2, solid_capstyle='round', linestyle=':')

ax3.set_xlabel('Federated Learning Round', fontsize=20)
ax3.set_ylabel('F1 Score over Total Round Time', fontsize=20)
ax3.tick_params(axis='both', labelsize=20)
#ax3.legend(fontsize=14)
ax3.set_xticks(rounds) 
plt.tight_layout()
plt.savefig('ClientSelector3.pdf')
plt.show()
```
