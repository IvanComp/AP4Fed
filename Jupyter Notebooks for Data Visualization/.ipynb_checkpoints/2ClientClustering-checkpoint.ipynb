```python
import matplotlib.pyplot as plt
import numpy as np

# Dati originali
rounds = np.arange(1, 11)  # 10 round

# Accuracy
accuracy_P1 = [0.25, 0.37, 0.43, 0.48, 0.52, 0.54, 0.55, 0.55, 0.58, 0.59]
accuracy_P1_prime = [0.22, 0.42, 0.42, 0.47, 0.50, 0.53, 0.54, 0.57, 0.57, 0.58]
accuracy_P1_excluded = [0.22, 0.42, 0.42, 0.47, 0.50, 0.53, 0.54, 0.57, 0.57, 0.58]

# Total Round Time
time_P1 = [445, 506, 344, 367, 469, 334, 509, 450, 475, 326]
time_P1_prime = [2027, 1686, 1941, 1822, 1781, 1798, 1780, 1865, 1805, 1851]
time_P1_excluded = [209, 214, 218, 206, 220, 225, 218, 211, 215, 217] 

# Calcolo Accuracy over Efficiency (Accuracy / Total Round Time)
efficiency_P1 = np.array(accuracy_P1) / np.array(time_P1)
efficiency_P1_prime = np.array(accuracy_P1_prime) / np.array(time_P1_prime)
efficiency_P1_excluded = np.array(accuracy_P1_excluded) / np.array(time_P1_excluded)

# Errori fittizi per ogni serie (puoi sostituirli con i tuoi dati di errore reali)
error_accuracy_P1 = np.random.uniform(0.02, 0.05, len(rounds))
error_accuracy_P1_prime = np.random.uniform(0.01, 0.03, len(rounds))
error_accuracy_P1_excluded = np.random.uniform(0.01, 0.03, len(rounds))

error_time_P1 = np.random.uniform(20, 50, len(rounds))
error_time_P1_prime = np.random.uniform(100, 150, len(rounds))
error_time_P1_excluded = np.random.uniform(5, 15, len(rounds))

# Colori pastello
color_p1 = (102/255, 178/255, 255/255, 0.6)  # Azzurro pastello
color_p1_prime = (255/255, 153/255, 153/255, 0.6)  # Rosso pastello
color_p1_excluded = (255/255, 255/255, 102/255, 0.6)  # Giallo pastello

# Creazione del grafico con 3 colonne
fig, axs = plt.subplots(1, 3, figsize=(18, 6))  # 1 riga, 3 colonne

# Grafico 1: Accuracy x Rounds con bande di errore
axs[0].plot(rounds, accuracy_P1, color=color_p1, label='4 High-S', marker='o', linewidth=2, solid_capstyle='round')
axs[0].fill_between(rounds, np.array(accuracy_P1) - error_accuracy_P1, np.array(accuracy_P1) + error_accuracy_P1, color=color_p1, alpha=0.3)

axs[0].plot(rounds, accuracy_P1_prime, color=color_p1_prime, label='3 High-S + 1 Low-S (Included)', marker='o', linewidth=2, solid_capstyle='round')
axs[0].fill_between(rounds, np.array(accuracy_P1_prime) - error_accuracy_P1_prime, np.array(accuracy_P1_prime) + error_accuracy_P1_prime, color=color_p1_prime, alpha=0.3)

axs[0].plot(rounds, accuracy_P1_excluded, color=color_p1_excluded, label='3 High-S + 1 Low-S (Excluded)', marker='o', linewidth=2, solid_capstyle='round')
axs[0].fill_between(rounds, np.array(accuracy_P1_excluded) - error_accuracy_P1_excluded, np.array(accuracy_P1_excluded) + error_accuracy_P1_excluded, color=color_p1_excluded, alpha=0.3)

axs[0].set_title('Accuracy x Rounds', fontsize=12, fontweight='bold')
axs[0].set_xlabel('Rounds', fontsize=10, fontweight='bold')
axs[0].set_ylabel('Accuracy (F1)', fontsize=10, fontweight='bold')
axs[0].legend(fontsize=10)
axs[0].set_xticks(rounds)  # Mostra tutti i round sull'asse X

# Grafico 2: Total Round Time x Rounds con bande di errore
axs[1].plot(rounds, time_P1, color=color_p1, label='4 High-S', marker='o', linewidth=2, solid_capstyle='round')
axs[1].fill_between(rounds, np.array(time_P1) - error_time_P1, np.array(time_P1) + error_time_P1, color=color_p1, alpha=0.3)

axs[1].plot(rounds, time_P1_prime, color=color_p1_prime, label='3 High-S + 1 Low-S (Included)', marker='o', linewidth=2, solid_capstyle='round')
axs[1].fill_between(rounds, np.array(time_P1_prime) - error_time_P1_prime, np.array(time_P1_prime) + error_time_P1_prime, color=color_p1_prime, alpha=0.3)

axs[1].plot(rounds, time_P1_excluded, color=color_p1_excluded, label='3 High-S + 1 Low-S (Excluded)', marker='o', linewidth=2, solid_capstyle='round')
axs[1].fill_between(rounds, np.array(time_P1_excluded) - error_time_P1_excluded, np.array(time_P1_excluded) + error_time_P1_excluded, color=color_p1_excluded, alpha=0.3)

axs[1].set_title('Total Round Time x Rounds', fontsize=12, fontweight='bold')
axs[1].set_xlabel('Rounds', fontsize=10, fontweight='bold')
axs[1].set_ylabel('Total Round Time (s)', fontsize=10, fontweight='bold')
axs[1].legend(fontsize=10)
axs[1].set_xticks(rounds)  # Mostra tutti i round sull'asse X

# Grafico 3: Accuracy over Efficiency con bande di errore (calcolato)
axs[2].plot(rounds, efficiency_P1, color=color_p1, label='4 High-S', marker='s', linewidth=2, solid_capstyle='round')
axs[2].plot(rounds, efficiency_P1_prime, color=color_p1_prime, label='3 High-S + 1 Low-S (Included)', marker='s', linewidth=2, solid_capstyle='round')
axs[2].plot(rounds, efficiency_P1_excluded, color=color_p1_excluded, label='3 High-S + 1 Low-S (Excluded)', marker='s', linewidth=2, solid_capstyle='round')

axs[2].set_title('Accuracy over Efficiency', fontsize=12, fontweight='bold')
axs[2].set_xlabel('Rounds', fontsize=10, fontweight='bold')
axs[2].set_ylabel('Accuracy / Total Round Time', fontsize=10, fontweight='bold')
axs[2].legend(fontsize=10)
axs[2].set_xticks(rounds)  # Mostra tutti i round sull'asse X

# Spaziatura tra i subplots
plt.tight_layout()
plt.savefig('LineGraphsWithErrorBands_Horizontal_Pastel_Rounded.pdf')
plt.show()
```
