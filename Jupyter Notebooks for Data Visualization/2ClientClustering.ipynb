```python
import matplotlib.pyplot as plt
import numpy as np

rounds = np.arange(1, 11)  

# Dati di accuracy per le diverse configurazioni
accuracy_4a4b_cluster = [0.07, 0.16, 0.24, 0.28, 0.31, 0.36, 0.38, 0.4, 0.43, 0.45]
accuracy_4a4b_no_cluster = [0.07, 0.08, 0.11, 0.19, 0.17, 0.21, 0.16, 0.22, 0.19, 0.2]
accuracy_5a3b_cluster = [0.08, 0.21, 0.28, 0.33, 0.35, 0.39, 0.4, 0.43, 0.45, 0.47]
accuracy_5a3b_no_cluster = [0.08, 0.05, 0.06, 0.13, 0.17, 0.16, 0.19, 0.17, 0.14, 0.19]
accuracy_6a2b_cluster = [0.06, 0.16, 0.23, 0.29, 0.34, 0.37, 0.4, 0.4, 0.44, 0.44]
accuracy_6a2b_no_cluster = [0.12, 0.14, 0.18, 0.21, 0.21, 0.2, 0.22, 0.21, 0.24, 0.26]

# Colori e stili
color_4a4b_cluster = '#2ca02c'  
color_4a4b_no_cluster = '#ff7f0e'
color_5a3b_cluster = '#1f77b4'  
color_5a3b_no_cluster = '#d62728'
color_6a2b_cluster = '#9467bd'
color_6a2b_no_cluster = '#8c564b'

fig, ax1 = plt.subplots(figsize=(10, 6))

# Aree di sfondo per le soglie
ax1.fill_between(rounds, 0.4, 0.5, color='lightgreen', alpha=0.2, label='With Cluster')
ax1.fill_between(rounds, 0, 0.26, color='lightcoral', alpha=0.2, label='Without Cluster')

# Plotta le curve
ax1.plot(rounds, accuracy_4a4b_cluster, color=color_4a4b_cluster, label='4a4b with cluster', marker='o', linewidth=2, solid_capstyle='round', linestyle='-')
ax1.plot(rounds, accuracy_4a4b_no_cluster, color=color_4a4b_no_cluster, label='4a4b no cluster', marker='s', linewidth=2, solid_capstyle='round', linestyle='--')
ax1.plot(rounds, accuracy_5a3b_cluster, color=color_5a3b_cluster, label='5a3b with cluster', marker='^', linewidth=2, solid_capstyle='round', linestyle=':')
ax1.plot(rounds, accuracy_5a3b_no_cluster, color=color_5a3b_no_cluster, label='5a3b no cluster', marker='d', linewidth=2, solid_capstyle='round', linestyle='-.')
ax1.plot(rounds, accuracy_6a2b_cluster, color=color_6a2b_cluster, label='6a2b with cluster', marker='*', linewidth=2, solid_capstyle='round', linestyle='-')
ax1.plot(rounds, accuracy_6a2b_no_cluster, color=color_6a2b_no_cluster, label='6a2b no cluster', marker='p', linewidth=2, solid_capstyle='round', linestyle='--')

# Linee di soglia
ax1.axhline(0.4, color='green', linestyle='--', linewidth=0.5)
ax1.axhline(0.26, color='red', linestyle='--', linewidth=0.5)

# Etichette, legenda e asse
ax1.set_xlabel('Federated Learning Round', fontsize=20)
ax1.set_ylabel('F1 Score', fontsize=20)
ax1.set_ylim(0, 0.5)
ax1.tick_params(axis='both', labelsize=20)
ax1.legend(fontsize=14, loc="upper left") 
ax1.set_xticks(rounds)

plt.tight_layout()
plt.savefig('EnhancedClientClustering1.pdf')
plt.show()
```
