```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from matplotlib import rcParams

plt.rcParams['font.family'] = 'CMU Serif'

rounds = np.arange(1, 11)  

# Accuracy (F1)
accuracy_6b = [0.38, 0.41, 0.43, 0.44, 0.45, 0.46, 0.46, 0.47, 0.46, 0.47]
accuracy_2a4b = [0.27, 0.40, 0.47, 0.50, 0.54, 0.56, 0.58, 0.60, 0.61, 0.61]
accuracy_3a3b = [0.30, 0.40, 0.46, 0.51, 0.54, 0.57, 0.58, 0.60, 0.61, 0.63]
accuracy_4a2b = [0.28, 0.40, 0.46, 0.52, 0.55, 0.57, 0.59, 0.61, 0.62, 0.63]
accuracy_6a = [0.25, 0.28, 0.33, 0.35, 0.35, 0.36, 0.37, 0.38, 0.37, 0.36]

# Total Round Time (s)
time_6b = [1074, 857, 916, 934, 955, 939, 779, 1009, 1001, 906]
time_2a4b = [820, 799, 813, 895, 781, 748, 852, 704, 777, 882]
time_3a3b = [588, 579, 585, 574, 599, 602, 541, 526, 592, 705]
time_4a2b = [758, 774, 790, 696, 805, 736, 758, 701, 768, 743]
time_6a = [527, 509, 600, 499, 541, 402, 497, 538, 568, 529]

# Accuracy over Efficiency
efficiency_6b = [0.35, 0.48, 0.47, 0.47, 0.47, 0.49, 0.59, 0.47, 0.46, 0.52]
efficiency_2a4b = [0.32, 0.50, 0.57, 0.56, 0.69, 0.75, 0.68, 0.85, 0.78, 0.69]
efficiency_3a3b = [0.50, 0.69, 0.79, 0.89, 0.89, 0.94, 1.07, 1.14, 1.03, 0.89]
efficiency_4a2b = [0.36, 0.52, 0.58, 0.74, 0.68, 0.77, 0.78, 0.86, 0.81, 0.85]
efficiency_6a = [0.47, 0.55, 0.55, 0.70, 0.65, 0.90, 0.74, 0.71, 0.65, 0.68]

color_6b = '#1f77b4'  
color_2a4b = '#d62728' 
color_3a3b = '#ff7f0e' 
color_4a2b = '#2ca02c'  
color_6a = '#9467bd'  

# Accuracy x Rounds
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.plot(rounds, accuracy_6b, color=color_6b, label='6b', marker='o', linewidth=2, solid_capstyle='round', linestyle='-')
ax1.plot(rounds, accuracy_2a4b, color=color_2a4b, label='2a4b', marker='s', linewidth=2, solid_capstyle='round', linestyle='--')
ax1.plot(rounds, accuracy_3a3b, color=color_3a3b, label='3a3b', marker='^', linewidth=2, solid_capstyle='round', linestyle=':')
ax1.plot(rounds, accuracy_4a2b, color=color_4a2b, label='4a2b', marker='d', linewidth=2, solid_capstyle='round', linestyle='-.')
ax1.plot(rounds, accuracy_6a, color=color_6a, label='6a', marker='*', linewidth=2, solid_capstyle='round', linestyle='-')
ax1.set_xlabel('Federated Learning Round', fontsize=20)
ax1.set_ylabel('FI Score', fontsize=20)
ax1.tick_params(axis='both', labelsize=20)
ax1.legend(fontsize=20)
ax1.set_xticks(rounds)
plt.tight_layout()
plt.savefig('ClientClustering1.pdf')
plt.show()

# Total Round Time x Rounds
fig, ax2 = plt.subplots(figsize=(10, 6))
ax2.plot(rounds, time_6b, color=color_6b, label='6b', marker='o', linewidth=2, solid_capstyle='round', linestyle='-')
ax2.plot(rounds, time_2a4b, color=color_2a4b, label='2a4b', marker='s', linewidth=2, solid_capstyle='round', linestyle='--')
ax2.plot(rounds, time_3a3b, color=color_3a3b, label='3a3b', marker='^', linewidth=2, solid_capstyle='round', linestyle=':')
ax2.plot(rounds, time_4a2b, color=color_4a2b, label='4a2b', marker='d', linewidth=2, solid_capstyle='round', linestyle='-.')
ax2.plot(rounds, time_6a, color=color_6a, label='6a', marker='*', linewidth=2, solid_capstyle='round', linestyle='-')
ax2.set_xlabel('Federated Learning Round', fontsize=20)
ax2.set_ylabel('Total Round Time (s)', fontsize=20)
ax2.tick_params(axis='both', labelsize=20)
#ax2.legend(fontsize=20)
ax2.set_xticks(rounds)
plt.tight_layout()
plt.savefig('ClientClustering2.pdf')
plt.show()

# Accuracy over Efficiency
fig, ax3 = plt.subplots(figsize=(10, 6))
ax3.plot(rounds, efficiency_6b, color=color_6b, label='6b', marker='o', linewidth=2, solid_capstyle='round', linestyle='-')
ax3.plot(rounds, efficiency_2a4b, color=color_2a4b, label='2a4b', marker='s', linewidth=2, solid_capstyle='round', linestyle='--')
ax3.plot(rounds, efficiency_3a3b, color=color_3a3b, label='3a3b', marker='^', linewidth=2, solid_capstyle='round', linestyle=':')
ax3.plot(rounds, efficiency_4a2b, color=color_4a2b, label='4a2b', marker='d', linewidth=2, solid_capstyle='round', linestyle='-.')
ax3.plot(rounds, efficiency_6a, color=color_6a, label='6a', marker='*', linewidth=2, solid_capstyle='round', linestyle='-')
ax3.set_xlabel('Federated Learning Round', fontsize=20)
ax3.set_ylabel('Accuracy / Total Round Time', fontsize=20)
ax3.tick_params(axis='both', labelsize=20)
#ax3.legend(fontsize=20)
ax3.set_xticks(rounds)
plt.tight_layout()
plt.savefig('ClientClustering3.pdf')
plt.show()

#DATASET PARTITION
rcParams['font.family'] = 'CMU Serif'

CLASS_NAMES = ['airplane', 'automobile', 'bird', 'cat', 'deer', 
               'dog', 'frog', 'horse', 'ship', 'truck']

clients_a_data_percent = {f"Client {i+1} - A": [10] * len(CLASS_NAMES) for i in range(4)}

total_samples = 50000
non_iid_data = {
    "Client 1 - B": [4720, 15248, 1714, 1978, 7463, 65, 7089, 2123, 4353, 5242],
    "Client 2 - B": [127, 450, 18256, 9645, 637, 1109, 7812, 2180, 664, 9115],
    "Client 3 - B": [3041, 233, 67, 13316, 5062, 12488, 1253, 3635, 1910, 8991],
    "Client 4 - B": [7505, 522, 6391, 10835, 1101, 5027, 512, 11256, 2196, 4650]
}
non_iid_data_percent = {
    client: [(val / total_samples) * 100 for val in data] 
    for client, data in non_iid_data.items()
}

fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True, gridspec_kw={'width_ratios': [1, 1]})

class_colors = sns.color_palette("RdYlGn", len(CLASS_NAMES))

bar_height = 0.9

for i, (client, data) in enumerate(reversed(clients_a_data_percent.items())):
    left = 0
    for j, percentage in enumerate(data):
        axes[0].barh(i, percentage, height=bar_height, left=left, color=class_colors[j], edgecolor='none')
        left += percentage
    axes[0].text(-5, i, client, va='center', ha='right', fontsize=20, color="black")

axes[0].set_xlim(0, 100)
axes[0].set_xticks([0, 25, 50, 75, 100])
axes[0].set_xlabel("Class Distribution (%)", fontsize=20)
#axes[0].set_ylabel("Client ID")
axes[0].tick_params(axis='both', labelsize=20)
axes[0].set_yticks(range(len(clients_a_data_percent)))
axes[0].set_yticklabels([])

for i, (client, data) in enumerate(reversed(non_iid_data_percent.items())):
    left = 0
    for j, percentage in enumerate(data):
        axes[1].barh(i, percentage, height=bar_height, left=left, color=class_colors[j], edgecolor='none')
        left += percentage
    axes[1].text(102, i, client, va='center', ha='left', fontsize=20, color="black")

axes[1].set_xlim(0, 100)
axes[1].set_xticks([0, 25, 50, 75, 100])
axes[1].set_xlabel("Class Distribution (%)", fontsize=20)
axes[1].tick_params(axis='both', labelsize=20)
axes[1].set_yticks(range(len(non_iid_data_percent)))
axes[1].set_yticklabels([])

fig.subplots_adjust(hspace=0.05)
fig.legend(
    CLASS_NAMES, loc="lower center", 
    bbox_to_anchor=(0.5, -0.16), ncol=5, fontsize=15
)

plt.tight_layout()
plt.savefig('DataPartition.pdf', bbox_inches='tight')
plt.show()
```
